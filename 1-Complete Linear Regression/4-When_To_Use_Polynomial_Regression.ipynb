# When to Use Polynomial Regression

Polynomial Regression is an extension of Linear Regression that models **non-linear relationships** between the independent variables and the dependent variable. Although the relationship between the variables is non-linear, the model remains **linear in its coefficients**.

## You should use Polynomial Regression when:

### 1. The relationship between the input and output is non-linear

* The target variable does not change at a constant rate with respect to the input.
* Instead, the data follows a curved trend.

**Example:**

* Student performance may improve as study hours increase, but after a certain point, excessive studying may lead to fatigue, causing performance to decline.

---

### 2. The data exhibits a curved pattern

* A scatter plot reveals that the observations follow a curved rather than a straight-line trend.
* Common curve shapes include:

  * U-shaped
  * Inverted U-shaped
  * S-shaped
  * Wave-like patterns

---

### 3. Linear Regression underfits the data

* If a Simple or Multiple Linear Regression model cannot adequately capture the relationship, Polynomial Regression can improve performance by introducing polynomial features.

---

### 4. You want to model feature interactions and higher-order effects

* Polynomial Regression allows the model to learn:

  * Squared terms ($x^2$)
  * Cubic terms ($x^3$)
  * Higher-order polynomial terms
  * Interaction terms between features

These additional features enable the model to capture more complex relationships.

---

## Real-World Applications

* Predicting plant growth over time.
* Modeling population growth.
* Predicting fuel efficiency based on vehicle speed.
* Modeling learning performance versus study time.
* Estimating product demand based on pricing strategies.
* Predicting temperature variation throughout the day.

---

# Mathematical Representation

## Case 1: One Independent Variable

### Degree 1 (Linear Regression)

$$
y=\beta_0+\beta_1x
$$

---

### Degree 2 (Quadratic Regression)

$$
y=\beta_0+\beta_1x+\beta_2x^2
$$

---

### Degree 3 (Cubic Regression)

$$
y=\beta_0+\beta_1x+\beta_2x^2+\beta_3x^3
$$

---

where

* $x$ = Independent variable
* $y$ = Dependent variable
* $\beta_0$ = Intercept
* $\beta_1,\beta_2,\beta_3$ = Model coefficients

---

## Case 2: Two Independent Variables

### Degree 1 (Multiple Linear Regression)

$$
y=\beta_0+\beta_1x_1+\beta_2x_2
$$

---

### Degree 2 (Polynomial Regression)

$$
y=\beta_0+\beta_1x_1+\beta_2x_2+\beta_3x_1^2+\beta_4x_2^2+\beta_5x_1x_2
$$

This model includes:

* Linear terms
* Squared terms
* Interaction term

---

## Feature Expansion

Polynomial Regression works by transforming the original features into polynomial features.

For example,

| Original Feature | Degree | Transformed Features             |
| ---------------- | ------ | -------------------------------- |
| $x$              | 2      | $x,;x^2$                         |
| $x$              | 3      | $x,;x^2,;x^3$                    |
| $x_1,x_2$        | 2      | $x_1,;x_2,;x_1^2,;x_2^2,;x_1x_2$ |

After feature transformation, **Linear Regression** is applied to these newly created features.

---

# When NOT to Use Polynomial Regression

❌ **The relationship is already linear**

* A simple linear model is more interpretable and computationally efficient.

---

❌ **The polynomial degree is too high**

* High-degree polynomials may fit the training data extremely well but perform poorly on unseen data due to overfitting.

---

❌ **The dataset contains excessive noise**

* Polynomial Regression may learn noise instead of the underlying relationship.

---

❌ **The dataset is very small**

* Higher-degree models require more data to estimate coefficients reliably.

---

# Advantages

* Captures non-linear relationships.
* Improves prediction accuracy for curved data.
* Easy to implement using feature transformation.
* Retains the simplicity of Linear Regression after feature expansion.

---

# Limitations

* Sensitive to outliers.
* High-degree models can overfit.
* Interpretation becomes more difficult as polynomial degree increases.
* Extrapolation outside the training range may produce unrealistic predictions.

---

# Summary

| Condition                        | Polynomial Regression? |
| -------------------------------- | ---------------------- |
| Non-linear relationship          | ✅ Yes                  |
| Curved data pattern              | ✅ Yes                  |
| Linear Regression underfits      | ✅ Yes                  |
| Higher-order relationships exist | ✅ Yes                  |
| Relationship is already linear   | ❌ No                   |
| Very high polynomial degree      | ❌ Risk of Overfitting  |

---

# Key Intuition

* **Degree 1** → Straight line (Linear Regression)
* **Degree 2** → One bend (Quadratic Curve)
* **Degree 3** → Two bends (Cubic Curve)
* **Higher Degree** → More flexible curve with increased risk of overfitting.

> **Rule of Thumb:** Polynomial Regression should be used when the relationship between the predictors and the target is non-linear, but can be approximated using polynomial functions. Choosing the appropriate polynomial degree is essential to achieve a balance between underfitting and overfitting.
