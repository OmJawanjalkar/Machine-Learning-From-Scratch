# Principal Component Analysis (PCA) - Top Interview Questions with Answers

---

# 1. What is Principal Component Analysis (PCA)?

### Answer

Principal Component Analysis (PCA) is an **unsupervised machine learning algorithm** used for **dimensionality reduction**. It transforms a dataset containing many correlated features into a new set of **uncorrelated variables** called **Principal Components**.

The main objective of PCA is to retain the maximum amount of information (variance) while reducing the number of features.

### Example

Suppose a dataset has 100 features.

Instead of training a model on all 100 features, PCA may reduce it to 10 principal components while preserving about 95% of the information.

---

# 2. Why do we use PCA?

### Answer

PCA is used to:

- Reduce dimensionality
- Remove redundant information
- Reduce computation time
- Reduce overfitting
- Remove multicollinearity
- Visualize high-dimensional data
- Improve model performance

---

# 3. What is a Principal Component?

### Answer

A Principal Component is a **new feature** created as a linear combination of the original features.

Each principal component satisfies two conditions:

- Captures the maximum possible variance.
- Is orthogonal (perpendicular) to all previous principal components.

Mathematically,

$$
PC_1 = w_1X_1 + w_2X_2 + w_3X_3 + \cdots + w_nX_n
$$

where

- $X_i$ = Original features
- $w_i$ = Weights (Eigenvectors)

---

# 4. What is Variance in PCA?

### Answer

Variance measures how much the data points are spread from the mean.

Higher variance means more information.

PCA tries to preserve the directions having the highest variance.

Formula

$$
Variance = \frac{1}{n}\sum_{i=1}^{n}(x_i-\mu)^2
$$

where

- $x_i$ = Data point
- $\mu$ = Mean
- $n$ = Number of observations

---

# 5. Why is variance important in PCA?

### Answer

PCA assumes that:

**Features with higher variance contain more useful information.**

Features with very low variance contribute less to explaining the dataset.

Therefore PCA keeps components with maximum variance.

---

# 6. Why should we standardize the data before PCA?

### Answer

PCA is sensitive to feature scale.

If one feature has values between 0–1 and another between 0–10000, the larger feature dominates the variance.

Therefore, features must be standardized.

Standardization formula

$$
Z=\frac{X-\mu}{\sigma}
$$

where

- $\mu$ = Mean
- $\sigma$ = Standard deviation

After scaling:

- Mean = 0
- Standard Deviation = 1

---

# 7. What is the Covariance Matrix?

### Answer

The covariance matrix measures how features vary together.

Positive covariance:
Both variables increase together.

Negative covariance:
One increases while the other decreases.

Zero covariance:
Variables are unrelated.

Covariance formula

$$
Cov(X,Y)=\frac{1}{n-1}\sum_{i=1}^{n}(X_i-\bar X)(Y_i-\bar Y)
$$

Covariance Matrix

$$
C=
\begin{bmatrix}
Cov(X_1,X_1) & Cov(X_1,X_2) \\
Cov(X_2,X_1) & Cov(X_2,X_2)
\end{bmatrix}
$$

---

# 8. What are Eigenvalues?

### Answer

Eigenvalues represent the amount of variance captured by each principal component.

A larger eigenvalue means the component contains more information.

Formula

$$
Av=\lambda v
$$

where

- $A$ = Covariance Matrix
- $v$ = Eigenvector
- $\lambda$ = Eigenvalue

---

# 9. What are Eigenvectors?

### Answer

Eigenvectors define the direction of the new principal components.

They determine how the original features are combined to form each principal component.

Each principal component is an eigenvector of the covariance matrix.

---

# 10. Explain the mathematical steps of PCA.

### Answer

The PCA algorithm follows these steps:

### Step 1

Standardize the data

$$
Z=\frac{X-\mu}{\sigma}
$$

↓

### Step 2

Compute the covariance matrix

$$
C=\frac{1}{n-1}X^TX
$$

↓

### Step 3

Find Eigenvalues and Eigenvectors

$$
Av=\lambda v
$$

↓

### Step 4

Sort eigenvalues from largest to smallest.

↓

### Step 5

Choose the top k principal components.

↓

### Step 6

Project the data

$$
Y=XW
$$

where

- $X$ = Original Data
- $W$ = Matrix of selected eigenvectors
- $Y$ = Reduced Dataset

---

# 11. How do you choose the number of Principal Components?

### Answer

The number of principal components is selected using the **Explained Variance Ratio**.

Formula

$$
Explained\ Variance\ Ratio=
\frac{\lambda_i}
{\sum_{j=1}^{n}\lambda_j}
$$

Cumulative Explained Variance

$$
\sum_{i=1}^{k}
\frac{\lambda_i}
{\sum_{j=1}^{n}\lambda_j}
\geq95\%
$$

Generally,

- 90%
- 95%
- 99%

variance is preserved.

---

# 12. What is Explained Variance?

### Answer

Explained variance indicates how much information is retained by each principal component.

Example

| Component | Explained Variance |
|-----------|-------------------|
| PC1 | 60% |
| PC2 | 25% |
| PC3 | 10% |
| PC4 | 5% |

Total variance = 100%

Keeping PC1 and PC2 preserves

$$
60\%+25\%=85\%
$$

of the information.

---

# 13. Why are Principal Components orthogonal?

### Answer

Orthogonal means perpendicular.

Their dot product equals zero.

$$
PC_i^TPC_j=0
$$

This ensures

- No redundancy
- No correlation
- Independent components

---

# 14. Does PCA perform feature selection?

### Answer

No.

PCA performs **feature extraction**, not feature selection.

Feature Selection:
Keeps original features.

Feature Extraction:
Creates completely new features (Principal Components).

---

# 15. Can PCA handle correlated features?

### Answer

Yes.

In fact, PCA works best when features are highly correlated.

It transforms correlated variables into uncorrelated principal components.

### Answer

| Covariance | Correlation |
|------------|-------------|
| Scale dependent | Scale independent |
| Value can be any number | Value lies between -1 and 1 |

# 17. What are the advantages of PCA?

### Answer

Advantages include:

- Reduces dimensionality
- Faster model training
- Removes multicollinearity
- Removes noise
- Reduces overfitting
- Improves visualization
- Compresses data

---

# 18. What are the disadvantages of PCA?

### Answer

Disadvantages include:

- Difficult to interpret principal components
- Information loss after reduction
- Sensitive to scaling
- Assumes linear relationships
- Not suitable for all datasets

---

# 19. When should you use PCA?

### Answer

Use PCA when:

- Dataset has many features
- Features are highly correlated
- Training is slow
- Need visualization
- Need compression
- Want to reduce overfitting

---

# 20. When should you NOT use PCA?

### Answer

Avoid PCA when:

- Features are already independent
- Interpretability is important
- Dataset has very few features
- Non-linear relationships dominate

|------|-------------------|

| No target variable | Requires target variable |
| Reduces dimensions | Predicts target |
| Maximizes variance | Minimizes prediction error |

# 23. What is the objective function of PCA?

### Answer

PCA aims to maximize the variance of the projected data.

Mathematically,

$$
\max_W Var(XW)
$$

subject to

$$
W^TW=I
$$

where

- $W$ = Projection Matrix
- $I$ = Identity Matrix
