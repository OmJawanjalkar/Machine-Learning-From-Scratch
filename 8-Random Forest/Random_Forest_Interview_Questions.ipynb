# 🎯 Top 20 Random Forest Interview Questions and Answers

## 1. What is Random Forest?

**Answer:**

Random Forest is a supervised machine learning algorithm that combines multiple Decision Trees to improve prediction accuracy and reduce overfitting. It can be used for both **classification** and **regression** problems.

---

## 2. Why is it called Random Forest?

**Answer:**

It is called a **Random Forest** because:

* It creates multiple Decision Trees (a forest of trees).
* Each tree is trained on a random subset of the training data.
* At every split, a random subset of features is selected.

This randomness makes the model more accurate and less prone to overfitting.

---

## 3. What is the difference between Decision Tree and Random Forest?

**Answer:**

| Decision Tree     | Random Forest          |
| ----------------- | ---------------------- |
| Single Tree       | Multiple Trees         |
| High Overfitting  | Less Overfitting       |
| High Variance     | Low Variance           |
| Faster            | Slightly Slower        |
| Easy to Interpret | Difficult to Interpret |

---

## 4. Why is Random Forest better than Decision Tree?

**Answer:**

Random Forest reduces overfitting by combining predictions from multiple Decision Trees. It provides higher accuracy, better generalization, and more stable predictions.

---

## 5. What is Ensemble Learning?

**Answer:**

Ensemble Learning is a technique where multiple machine learning models are combined to produce a better prediction than a single model.

Random Forest is an example of an **Ensemble Learning** algorithm.

---

## 6. What is Bagging?

**Answer:**

Bagging (Bootstrap Aggregating) is an ensemble technique where multiple models are trained independently on different bootstrap samples, and their predictions are combined.

Random Forest uses the **Bagging** technique.

---

## 7. What is Bootstrap Sampling?

**Answer:**

Bootstrap Sampling is the process of randomly selecting rows from the dataset **with replacement**.

This means:

* Some rows may appear multiple times.
* Some rows may not appear at all.

---

## 8. What is Feature Sampling?

**Answer:**

Feature Sampling means selecting a random subset of features at every split while building each Decision Tree.

This increases diversity among the trees and reduces correlation.

---

## 9. Why does Random Forest use Random Features?

**Answer:**

Random feature selection prevents all Decision Trees from becoming identical, improves diversity, reduces correlation, and increases prediction accuracy.

---

## 10. What is Majority Voting?

**Answer:**

Majority Voting is used in **Random Forest Classification**.

Each Decision Tree predicts a class, and the class receiving the highest number of votes becomes the final prediction.

---

## 11. What is Averaging in Random Forest?

**Answer:**

Averaging is used in **Random Forest Regression**.

The final prediction is calculated by taking the average of the predictions from all Decision Trees.

---

## 12. What is Out-of-Bag (OOB) Error?

**Answer:**

Out-of-Bag (OOB) Error is an internal validation method.

The samples not selected during bootstrap sampling are used to evaluate the model without needing a separate validation dataset.

---

## 13. Can Random Forest handle missing values?

**Answer:**

Yes. Random Forest can handle missing values better than many algorithms, although preprocessing the data is still recommended for optimal performance.

---

## 14. Can Random Forest handle categorical and numerical data?

**Answer:**

Yes.

Random Forest can work with both categorical and numerical features after appropriate preprocessing, such as encoding categorical variables.

---

## 15. What are the important hyperparameters in Random Forest?

**Answer:**

Important hyperparameters include:

* n_estimators
* max_depth
* max_features
* min_samples_split
* min_samples_leaf
* criterion
* bootstrap
* random_state
* oob_score
* n_jobs

---

## 16. What are the advantages of Random Forest?

**Answer:**

* High prediction accuracy
* Reduces overfitting
* Handles large datasets
* Works with high-dimensional data
* Robust to noise and outliers
* Provides feature importance
* Suitable for classification and regression

---

## 17. What are the disadvantages of Random Forest?

**Answer:**

* Computationally expensive
* Requires more memory
* Slower training and prediction
* Difficult to interpret
* Large model size

---

## 18. When should you use Random Forest Classification?

**Answer:**

Use Random Forest Classification when the target variable is **categorical**.

Examples:

* Spam Detection
* Fraud Detection
* Customer Churn Prediction
* Disease Prediction
* Loan Approval

---

## 19. When should you use Random Forest Regression?

**Answer:**

Use Random Forest Regression when the target variable is **continuous (numerical)**.

Examples:

* House Price Prediction
* Salary Prediction
* Sales Forecasting
* Temperature Prediction
* Stock Price Prediction

---

## 20. Why is Random Forest called an Ensemble Algorithm?

**Answer:**

Random Forest is called an Ensemble Algorithm because it combines the predictions of multiple Decision Trees to produce a single, more accurate prediction.

Instead of relying on one model, it aggregates the outputs of many models, making it more robust and reliable.

---

# ⭐ Bonus Interview Question

## Explain the working of Random Forest in one minute.

**Answer:**

Random Forest is a supervised machine learning algorithm that builds multiple Decision Trees using different bootstrap samples of the training data. At each split, it considers a random subset of features, making each tree different. During prediction, it combines the outputs of all trees using **majority voting** for classification or **averaging** for regression. This approach reduces overfitting, improves accuracy, and provides better generalization than a single Decision Tree.
