# Top 10 KNN Interview Questions and Answers

## 1. What is K-Nearest Neighbors (KNN)?

**Answer:**

K-Nearest Neighbors (KNN) is a supervised machine learning algorithm used for both **classification** and **regression**. It predicts the output of a new data point by finding the **K nearest neighbors** in the training dataset.

* Classification → Majority Voting
* Regression → Average of Neighbor Values

---

## 2. Why is KNN called a Lazy Learning Algorithm?

**Answer:**

KNN is called a **Lazy Learning Algorithm** because it does not build a model during training. Instead, it stores the entire training dataset and performs all computations only when making predictions.

**Training Time:** Very Fast

**Prediction Time:** Slow

---

## 3. Why is KNN called a Non-Parametric Algorithm?

**Answer:**

KNN is non-parametric because it does not assume any underlying distribution of the data and does not learn fixed parameters like weights or coefficients.

---

## 4. Why is Feature Scaling Important in KNN?

**Answer:**

KNN uses distance calculations to find the nearest neighbors.

If one feature has much larger values than others, it dominates the distance calculation and leads to incorrect predictions.

Common scaling techniques:

* StandardScaler
* MinMaxScaler
* RobustScaler

---

## 5. How do you choose the value of K?

**Answer:**

The value of **K** is selected using **Cross-Validation**.

* Small K → Overfitting (High Variance)
* Large K → Underfitting (High Bias)

A common starting point is **K = 5**.

---

## 6. What distance metrics can be used in KNN?

**Answer:**

Common distance metrics include:

* Euclidean Distance
* Manhattan Distance
* Minkowski Distance
* Chebyshev Distance

Euclidean Distance is the default and most commonly used metric.

---

## 7. What is the Time Complexity of KNN?

**Answer:**

**Training Complexity:** O(1)

KNN does not train a model; it simply stores the data.

**Prediction Complexity:** O(n × d)

Where:

* n = Number of training samples
* d = Number of features

Prediction becomes slower as the dataset size increases.

---

## 8. What are the Advantages and Disadvantages of KNN?

### Advantages

* Simple to understand and implement.
* No training phase.
* Works for both classification and regression.
* Can model non-linear decision boundaries.

### Disadvantages

* Slow prediction on large datasets.
* Sensitive to feature scaling.
* Sensitive to noisy data and outliers.
* Suffers from the Curse of Dimensionality.

---

## 9. What is the Curse of Dimensionality?

**Answer:**

As the number of features (dimensions) increases, the distance between data points becomes less meaningful. This makes it difficult for KNN to identify the true nearest neighbors, reducing its performance.

---

## 10. What is the difference between KNN Classifier and KNN Regressor?

| KNN Classifier          | KNN Regressor                   |
| ----------------------- | ------------------------------- |
| Used for Classification | Used for Regression             |
| Predicts Class Labels   | Predicts Numerical Values       |
| Uses Majority Voting    | Uses Average of Neighbor Values |
| Output is Categorical   | Output is Continuous            |

---

# Bonus Interview Questions

### 11. Why is KNN sensitive to outliers?

Outliers can become one of the nearest neighbors and significantly influence the prediction, especially when using a small value of K.

---

### 12. Why does KNN perform poorly on large datasets?

Because it computes the distance between the query point and every training sample during prediction, making it computationally expensive.

---

### 13. What happens if K = 1?

The model becomes highly sensitive to noise and may overfit the training data.

---

### 14. What happens if K is too large?

The model becomes too generalized, may ignore local patterns, and can underfit the data.

---

### 15. How can you make KNN faster?

* Use **KD-Tree** for low-dimensional data.
* Use **Ball Tree** for high-dimensional data.
* Reduce the number of features using feature selection or dimensionality reduction.
* Use approximate nearest neighbor search for very large datasets.

---

# Interview Tips

* **Classification → Majority Voting**
* **Regression → Average of Neighbors**
* **Training Complexity → O(1)**
* **Prediction Complexity → O(n × d)**
* **Feature Scaling is Mandatory**
* **Small K → Overfitting**
* **Large K → Underfitting**
* **KD-Tree → Low-dimensional Data**
* **Ball Tree → High-dimensional Data**
* **Cross-Validation → Best method to choose K**
