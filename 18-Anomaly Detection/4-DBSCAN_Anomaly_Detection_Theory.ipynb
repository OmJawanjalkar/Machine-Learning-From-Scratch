# DBSCAN for Anomaly Detection

## What is DBSCAN?

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** is an **unsupervised machine learning algorithm** that groups together data points that are **closely packed (high density)** and labels points that are far away from any cluster as **noise (outliers/anomalies)**.

Unlike many clustering algorithms, DBSCAN can **detect anomalies automatically** because it identifies isolated points as **Noise**.

---

# Simple Definition

Imagine a crowded classroom where students are sitting in groups.

* Students sitting together form **clusters**.
* A student sitting alone in the corner is considered an **anomaly (noise point)**.

DBSCAN works in exactly the same way—it groups nearby data points and labels isolated points as anomalies.

---

# Types of Points in DBSCAN

DBSCAN classifies every data point into **three types**:

### 1. Core Point

A point that has at least **MinPts** neighboring points within a distance **ε (epsilon)**.

* Starts a new cluster.
* Represents a dense region.

### 2. Border Point

A point that has fewer than **MinPts** neighbors but lies within the neighborhood of a Core Point.

* Belongs to an existing cluster.
* Cannot start a new cluster.

### 3. Noise Point (Anomaly)

A point that is not connected to any cluster.

* Lies far away from dense regions.
* Considered an **anomaly (outlier)**.

---

# How DBSCAN Detects Anomalies

### Step 1

Choose two parameters:

* **ε (Epsilon):** Maximum distance between neighboring points.
* **MinPts:** Minimum number of nearby points required to form a cluster.

### Step 2

Pick any unvisited point.

### Step 3

Count how many neighboring points lie within the ε radius.

### Step 4

If the number of neighbors is **greater than or equal to MinPts**, the point becomes a **Core Point** and a new cluster is created.

### Step 5

Expand the cluster by adding all density-connected points.

### Step 6

If a point does **not** have enough neighbors and is **not connected to any Core Point**, it is labeled as **Noise (-1)**.

### Step 7

Repeat the process until every point has been visited.

---

# How Anomalies are Detected

After clustering:

* **Clustered Points → Normal Data**
* **Noise Points (-1) → Anomalies (Outliers)**

Therefore,

> **Any point that does not belong to a dense cluster is automatically detected as an anomaly.**

---

# Example

Suppose we have customer transaction data.

```
● ● ● ● ●

 ● ● ● ●

            ×
```

* **●** = Normal Customers
* **×** = Anomalous Customer (Noise)

The isolated point (**×**) has very few neighbors, so DBSCAN labels it as **Noise (-1)**.

---

# Advantages

* Detects anomalies automatically.
* No need to specify the number of clusters.
* Works well with irregular-shaped clusters.
* Robust to noisy data.
* Finds clusters based on data density.

---

# Disadvantages

* Choosing the correct **ε** and **MinPts** is important.
* Performs poorly when clusters have different densities.
* Less effective in high-dimensional datasets.

# Key Points to Remember

* DBSCAN is an **unsupervised clustering algorithm**.
* It groups **high-density** data points into clusters.
* Points that do **not** belong to any cluster are labeled as **Noise (-1)**.
* These **Noise Points** are treated as **Anomalies (Outliers)**.
* DBSCAN does **not** require the number of clusters beforehand, making it very useful for anomaly detection tasks.
