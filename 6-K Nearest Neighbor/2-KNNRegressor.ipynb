## KNN Regressor

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=1000, n_features=2, noise=10, random_state=42)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.33, random_state=42 )
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knnr = KNeighborsRegressor(n_neighbors=6, algorithm='auto')
knnr.fit(X_train,y_train)
y_pred = knnr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error

print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

## Hyperparameter Tuning with RandomizedSearchCV

```python
# Lets do some hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
n_neighbors=[2,3,4,5,6,7,8,9,10]
algorithm=['auto','ball_tree','kd_tree','brute']
weights=['uniform','distance']
param_grid={'n_neighbors':n_neighbors,'algorithm':algorithm,'weights':weights}

random_search=RandomizedSearchCV(KNeighborsRegressor(),param_distributions=param_grid,n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=-1)
random_search.fit(X_train,y_train)
random_search.best_params_
```

```python
y_pred=random_search.predict(X_test)
```

```python
# calculate metrics
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
