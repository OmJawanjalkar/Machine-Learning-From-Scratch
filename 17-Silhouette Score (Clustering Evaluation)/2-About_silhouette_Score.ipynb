# Silhouette Score

## What is Silhouette Score?

**Silhouette Score** is a metric used to measure **how good your clustering is**.

It checks two things:

- **Cohesion:** Are the points close to other points in the same cluster?
- **Separation:** Are the clusters far away from each other?

A good clustering has:
- Small distance within the same cluster.
- Large distance between different clusters.

---

## Formula

For each data point:

$$
S(i)=\frac{b(i)-a(i)}{\max(a(i),b(i))}
$$

Where:

- **\(a(i)\)** = Average distance to points in the **same cluster**.
- **\(b(i)\)** = Average distance to points in the **nearest neighboring cluster**.

---

## Score Range

$$
-1 \le S(i) \le 1
$$

| Score | Meaning |
|-------:|---------|
| **1** | Perfect clustering |
| **0.5 to 1** | Good clustering |
| **0** | Clusters overlap |
| **Less than 0** | Wrong cluster assignment |

---

## Example

Suppose:

$$
a = 2,\qquad b = 8
$$

Then,

$$
S=\frac{8-2}{8}=0.75
$$

**Silhouette Score = 0.75**

This means the clusters are **well separated**.

---

## Why Do We Use Silhouette Score?

- Evaluate clustering quality.
- Compare different clustering algorithms.
- Find the best number of clusters (**K**).
- Detect overlapping or poorly formed clusters.

---

## Simple Interview Answer

**Silhouette Score** is an internal clustering evaluation metric that measures how well data points fit within their own cluster compared to the nearest neighboring cluster. Its value ranges from **-1 to 1**, where a higher score indicates better clustering.
