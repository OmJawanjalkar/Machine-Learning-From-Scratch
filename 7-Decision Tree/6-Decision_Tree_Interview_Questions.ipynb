# Top 20 Decision Tree Interview Questions and Answers

## 1. What is a Decision Tree?

**Answer:**

A Decision Tree is a supervised machine learning algorithm used for both **classification** and **regression** tasks. It learns a series of **if-then rules** from the training data and uses them to make predictions.

---

## 2. What are the components of a Decision Tree?

**Answer:**

A Decision Tree consists of:

* Root Node
* Decision Node
* Leaf Node
* Branch
* Parent Node
* Child Node

---

## 3. What is the difference between a Decision Tree Classifier and a Decision Tree Regressor?

**Answer:**

| Decision Tree Classifier            | Decision Tree Regressor                             |
| ----------------------------------- | --------------------------------------------------- |
| Predicts categories or class labels | Predicts continuous numerical values                |
| Uses Entropy or Gini Index          | Uses Variance Reduction or Mean Squared Error (MSE) |
| Example: Spam or Not Spam           | Example: House Price Prediction                     |

---

## 4. What is a Root Node?

**Answer:**

The Root Node is the top-most node of the Decision Tree. It represents the entire dataset and is the first feature selected for splitting.

---

## 5. What is a Leaf Node?

**Answer:**

A Leaf Node is the final node of the Decision Tree where no further splitting occurs. It contains the final prediction.

---

## 6. What is a Pure Node?

**Answer:**

A Pure Node contains data points belonging to only one class.

Example:

* Pass
* Pass
* Pass

Since all records belong to the same class, the node is pure.

---

## 7. What is an Impure Node?

**Answer:**

An Impure Node contains data points from multiple classes.

Example:

* Pass
* Fail
* Pass
* Fail

Since multiple classes are present, the node is impure.

---

## 8. What is Entropy?

**Answer:**

Entropy measures the uncertainty or impurity in a dataset.

* Entropy = 0 → Completely Pure
* Entropy = 1 → Completely Mixed (Binary Classification)

The Decision Tree tries to minimize entropy after every split.

---

## 9. What is the Gini Index?

**Answer:**

The Gini Index measures the impurity of a node.

* Gini = 0 → Pure Node
* Gini = 0.5 → Maximum Impurity (Binary Classification)

The split with the lowest Gini Index is preferred.

---

## 10. What is Information Gain?

**Answer:**

Information Gain measures how much impurity decreases after a split.

The feature with the highest Information Gain is selected for splitting.

---

## 11. How does a Decision Tree choose the best feature?

**Answer:**

The algorithm evaluates all available features and selects the one that provides the maximum reduction in impurity.

For Classification:

* Highest Information Gain
* Lowest Gini Index

For Regression:

* Highest Variance Reduction

---

## 12. What is Variance Reduction?

**Answer:**

Variance Reduction is used in Decision Tree Regression.

It measures how much the variance decreases after splitting the data.

The feature with the highest variance reduction is selected.

---

## 13. Why is Feature Scaling not required for Decision Trees?

**Answer:**

Decision Trees split data based on feature values rather than distances.

Therefore, techniques such as Standardization and Normalization are not required.

---

## 14. What is Overfitting in a Decision Tree?

**Answer:**

Overfitting occurs when the tree becomes too deep and learns noise from the training data instead of general patterns.

As a result:

* High Training Accuracy
* Low Testing Accuracy

---

## 15. How can Overfitting be prevented?

**Answer:**

Overfitting can be reduced by:

* Pre-Pruning
* Post-Pruning
* Limiting Maximum Depth
* Increasing Minimum Samples Split
* Increasing Minimum Samples Leaf

---

## 16. What is Pre-Pruning?

**Answer:**

Pre-Pruning stops the growth of the tree while it is being built.

Common stopping conditions include:

* Maximum Depth
* Minimum Samples Split
* Minimum Samples Leaf

---

## 17. What is Post-Pruning?

**Answer:**

Post-Pruning first builds the complete tree and then removes unnecessary branches that do not improve model performance.

---

## 18. What are the advantages of Decision Trees?

**Answer:**

* Easy to understand and interpret
* Easy to visualize
* Handles both numerical and categorical data
* No feature scaling required
* Captures non-linear relationships
* Performs automatic feature selection
* Fast prediction

---

## 19. What are the disadvantages of Decision Trees?

**Answer:**

* Can easily overfit
* Sensitive to small changes in data
* Less accurate than ensemble methods
* Deep trees become difficult to interpret
* Can be biased toward features with many unique values

---

## 20. What is the difference between a Decision Tree and a Random Forest?

**Answer:**

| Decision Tree            | Random Forest                |
| ------------------------ | ---------------------------- |
| Single Tree              | Collection of Multiple Trees |
| Faster Training          | Slower Training              |
| Easy to Interpret        | Difficult to Interpret       |
| High Risk of Overfitting | Lower Risk of Overfitting    |
| Lower Accuracy           | Higher Accuracy              |
| Less Robust              | More Robust                  |

---

### 22. Why is Gini preferred over Entropy?

**Answer:**

Gini Index is computationally faster because it does not require logarithmic calculations. Therefore, many libraries such as Scikit-learn use Gini as the default criterion.

---

### 23. Why is a Decision Tree called a non-parametric algorithm?

**Answer:**

Decision Trees do not assume any fixed mathematical relationship or probability distribution between the input features and the target variable.

---

### 24. What hyperparameters can be tuned in a Decision Tree?

**Answer:**

* criterion
* max_depth
* min_samples_split
* min_samples_leaf
* max_features
* max_leaf_nodes
* splitter
* ccp_alpha

---

### 25. What are the stopping criteria in a Decision Tree?

**Answer:**

A Decision Tree stops splitting when:

* All samples belong to the same class.
* Maximum depth is reached.
* Minimum samples required for splitting are not available.
* Information Gain becomes very small.
* Minimum samples per leaf condition is met.
