# When to Use Logistic Regression

Logistic Regression is a **supervised machine learning algorithm** used for **classification problems**. It predicts the probability that an input belongs to a particular class.

## Use Logistic Regression When

* The target variable is **categorical** (e.g., Yes/No, 0/1, True/False).
* You need a **binary classification** model.
* The relationship between features and the target is approximately linear.
* The dataset is small to medium in size.
* You need a model that is **fast, simple, and easy to interpret**.
* You want to predict probabilities along with class labels.

### Examples

* Spam Email Detection (Spam / Not Spam)
* Diabetes Prediction (Diabetic / Not Diabetic)
* Customer Churn Prediction (Leave / Stay)
* Loan Approval (Approved / Rejected)
* Fraud Detection (Fraud / Not Fraud)

---

# When NOT to Use Logistic Regression

* The target variable is **continuous** (use Linear Regression instead).
* The data has a **highly complex or non-linear relationship**.
* There are many complex feature interactions that Logistic Regression cannot capture.
* The dataset is extremely large and another model performs better.
* You need the highest possible accuracy on complex datasets (Decision Trees, Random Forest, XGBoost, or Neural Networks may perform better).

---

# Advantages

* Simple and easy to understand.
* Fast to train.
* Works well for binary classification.
* Produces probability scores.
* Less prone to overfitting with regularization.

---

# Disadvantages

* Cannot capture complex non-linear relationships.
* Sensitive to outliers.
* Performance decreases if features are highly correlated.
* Not suitable for regression problems.

---

# Interview Tip

**Question:** When would you choose Logistic Regression?

**Answer:**
"I use Logistic Regression when the target variable is categorical, especially for binary classification problems. It is fast, easy to interpret, provides probability predictions, and works well when the relationship between the features and the target is approximately linear."
