# XGBoost Regression - Introduction & Step-by-Step Working

# 1. Introduction to XGBoost Regression

**XGBoost Regression** is a supervised machine learning algorithm used to predict **continuous numerical values**. It is an advanced implementation of the Gradient Boosting algorithm that builds multiple decision trees sequentially. Each new tree learns from the errors (residuals) made by the previous trees, gradually improving the prediction accuracy.

Unlike XGBoost Classification, which predicts probabilities, XGBoost Regression predicts **continuous values** such as prices, salaries, sales, or temperatures.

### Common Applications

* House Price Prediction
* Sales Forecasting
* Stock Price Prediction
* Temperature Prediction
* Energy Consumption Forecasting
* Salary Prediction

---

# 2. Workflow of XGBoost Regression

```text
Training Data
      │
      ▼
Initial Prediction
      │
      ▼
Calculate Residuals
      │
      ▼
Build Decision Tree
      │
      ▼
Calculate Similarity Score
      │
      ▼
Calculate Gain
      │
      ▼
Compute Leaf Weight
      │
      ▼
Update Prediction
      │
      ▼
Repeat Until All Trees Are Built
      │
      ▼
Final Prediction
```

---

# Step 1: Initialize the Prediction

XGBoost Regression starts by predicting the **mean of the target values**.

The initial prediction is

$$
\hat{y}_0=\frac{\sum y_i}{n}
$$

where,

* $\hat{y}_0$ = Initial Prediction
* $y_i$ = Actual Target Value
* $n$ = Number of Samples

### Example

Target Values:

$$
100,;200,;300
$$

Initial Prediction

$$
\hat{y}_0=\frac{100+200+300}{3}=200
$$

Every sample is initially predicted as **200**.

---

# Step 2: Calculate Residuals

Residuals represent the prediction error.

$$
\text{Residual}
===============

y-\hat{y}
$$

where,

* $y$ = Actual Value
* $\hat{y}$ = Predicted Value

### Example

| Actual | Prediction | Residual |
| -----: | ---------: | -------: |
|    100 |        200 |     -100 |
|    200 |        200 |        0 |
|    300 |        200 |      100 |

The next tree learns these residuals.

---

# Step 3: Build the First Decision Tree

The first decision tree is built using the residuals.

Example:

```text
          Size < 1500?
         /            \
     Leaf 1         Leaf 2
```

The tree tries to group similar residuals together.

---

# Step 4: Calculate Similarity Score

Each leaf receives a Similarity Score.

$$
\text{Similarity}
=================

\frac{\left(\sum g_i\right)^2}
{\sum h_i+\lambda}
$$

where,

* $g_i$ = Gradient
* $h_i$ = Hessian
* $\lambda$ = L2 Regularization

A higher Similarity Score indicates a better leaf.

---

# Step 5: Calculate Gain

Every possible split is evaluated.

$$
\text{Gain}
===========

\frac{1}{2}
\left(
\text{Left}
+
\text{Right}
------------

\text{Parent}
\right)
-\gamma
$$

where,

* Left = Left Similarity
* Right = Right Similarity
* Parent = Parent Similarity
* $\gamma$ = Minimum Loss Reduction

The split with the **highest Gain** is selected.

---

# Step 6: Calculate Leaf Weight

Each leaf predicts an output value.

$$
w
=

-\frac{\sum g_i}
{\sum h_i+\lambda}
$$

where,

* $w$ = Leaf Weight
* $\sum g_i$ = Sum of Gradients
* $\sum h_i$ = Sum of Hessians

This leaf weight represents the correction that will be added to the prediction.

---

# Step 7: Update the Prediction

The prediction is updated using the learning rate.

$$
\hat{y}_{new}
=============

\hat{y}_{old}
+
\eta \times w
$$

where,

* $\eta$ = Learning Rate
* $w$ = Leaf Weight

### Example

Old Prediction

$$
200
$$

Leaf Weight

$$
40
$$

Learning Rate

$$
0.3
$$

Updated Prediction

$$
200+(0.3\times40)=212
$$

The prediction becomes **212**, which is closer to the actual value.

---

# Step 8: Repeat the Process

The following steps are repeated until the stopping criteria are met:

1. Calculate Residuals
2. Build a Decision Tree
3. Calculate Similarity Score
4. Calculate Gain
5. Compute Leaf Weight
6. Update Prediction
7. Repeat

The process stops when:

* Maximum number of trees (**n_estimators**) is reached.
* The loss no longer decreases.
* Early stopping is triggered.

---

# Final Workflow

```text
Training Data
      │
      ▼
Initial Prediction (Mean)
      │
      ▼
Calculate Residuals
      │
      ▼
Build Decision Tree
      │
      ▼
Calculate Similarity Score
      │
      ▼
Calculate Gain
      │
      ▼
Compute Leaf Weight
      │
      ▼
Update Prediction
      │
      ▼
Repeat Until Error Becomes Small
      │
      ▼
Final Prediction
```

# Key Difference Between Classification and Regression

| XGBoost Classification                    | XGBoost Regression                         |
| ----------------------------------------- | ------------------------------------------ |
| Predicts class labels or probabilities    | Predicts continuous numerical values       |
| Initial prediction = Probability (0.5)    | Initial prediction = Mean of target values |
| Uses Log-Odds and Sigmoid Function        | Predicts values directly                   |
| Residual = Actual − Predicted Probability | Residual = Actual − Predicted Value        |
| Output = Class Prediction                 | Output = Numerical Prediction              |
