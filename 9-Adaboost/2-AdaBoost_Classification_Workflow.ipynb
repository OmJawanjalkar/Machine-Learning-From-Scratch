# AdaBoost Classification with Two Weak Learners

## 🎯 Example: Classifying Spam Emails Using Two Weak Learners

Suppose we want to classify whether an email is **Spam (1)** or **Not Spam (0)** using the AdaBoost algorithm.

### Dataset

| Email |   True Label |
| ----- | -----------: |
| A     | 0 (Not Spam) |
| B     | 0 (Not Spam) |
| C     |     1 (Spam) |
| D     |     1 (Spam) |

---

## Step 1: Initialize Equal Weights

Initially, AdaBoost assigns **equal importance (weight)** to every training sample because no model has been trained yet.

Since there are **4 emails**, the weight assigned to each email is:

[
\text{Weight} = \frac{1}{4} = 0.25
]

| Email | Initial Weight |
| ----- | -------------: |
| A     |           0.25 |
| B     |           0.25 |
| C     |           0.25 |
| D     |           0.25 |

---

## Step 2: Train the First Weak Learner

The first weak learner (typically a Decision Stump) is trained using the original dataset.

Suppose it makes the following predictions:

| Email | True Label | Prediction | Result    |
| ----- | ---------: | ---------: | --------- |
| A     |          0 |          0 | ✅ Correct |
| B     |          0 |          1 | ❌ Wrong   |
| C     |          1 |          1 | ✅ Correct |
| D     |          1 |          0 | ❌ Wrong   |

The first weak learner correctly classifies **A** and **C**, but misclassifies **B** and **D**.

Since **B** and **D** were classified incorrectly, AdaBoost **increases their weights** so that the next learner pays more attention to these difficult examples.

At the same time, the weights of correctly classified samples (**A** and **C**) are reduced.

---

## Step 3: Train the Second Weak Learner

The second weak learner is trained using the **updated sample weights**.

Because **B** and **D** now have higher weights, the model focuses more on classifying these samples correctly.

Suppose the second weak learner makes the following predictions:

| Email | True Label | Prediction | Result    |
| ----- | ---------: | ---------: | --------- |
| A     |          0 |          0 | ✅ Correct |
| B     |          0 |          0 | ✅ Correct |
| C     |          1 |          1 | ✅ Correct |
| D     |          1 |          1 | ✅ Correct |

The second learner successfully classifies **all four emails correctly**.

---

## Step 4: Combine the Weak Learners

AdaBoost does not rely on only one weak learner.

Instead, it combines the predictions of all weak learners using **weighted majority voting**.

* **Weak Learner 1** made more mistakes, so it receives a **lower voting weight**.
* **Weak Learner 2** performed better, so it receives a **higher voting weight**.

The final prediction is determined by combining the weighted votes from both learners.

---

## Final Prediction

After combining both weak learners, the final predictions are:

| Email | Final Prediction |
| ----- | ---------------: |
| A     |     Not Spam (0) |
| B     |     Not Spam (0) |
| C     |         Spam (1) |
| D     |         Spam (1) |

✅ **All four emails are classified correctly.**

---

## 🔑 Key Takeaways

* The first weak learner made mistakes while classifying **B** and **D**.
* AdaBoost increased the importance (weights) of these misclassified samples.
* The second weak learner focused more on these difficult examples and corrected the mistakes.
* Finally, AdaBoost combined the predictions of both weak learners using **weighted majority voting**.
* By combining several weak learners, AdaBoost creates a **strong classifier** with higher prediction accuracy.
