# XGBoost Classification – Step-by-Step Working

## Step 1: Initialize the Prediction

XGBoost starts by assigning an initial probability of **0.5** to every training sample.

Convert this probability into **Log-Odds** using:

$$
\text{Log-Odds}=\log\left(\frac{P}{1-P}\right)
$$

For,

$$
P=0.5
$$

$$
\text{Log-Odds}=0
$$

So, the initial prediction for every sample is **0** (Log-Odds).

---

## Step 2: Convert Log-Odds to Probability

Use the **Sigmoid Function** to convert Log-Odds into probability.

$$
P=\frac{1}{1+e^{-z}}
$$

where,

* $z$ = Log-Odds
* $P$ = Predicted Probability

Example:

If,

$$
z=0
$$

then,

$$
P=0.5
$$

---

## Step 3: Calculate Gradient (Residual)

The Gradient measures the prediction error.

$$
g_i=y_i-\hat{P}_i
$$

where,

* $y_i$ = Actual Label
* $\hat{P}_i$ = Predicted Probability

Example:

| Actual | Predicted | Gradient |
| -----: | --------: | -------: |
|      1 |       0.5 |      0.5 |
|      0 |       0.5 |     -0.5 |
|      1 |       0.5 |      0.5 |

---

## Step 4: Calculate Hessian

The Hessian is the second derivative of the loss function.

$$
h_i=\hat{P}_i(1-\hat{P}_i)
$$

Example:

If,

$$
\hat{P}=0.5
$$

then,

$$
h=0.5(1-0.5)=0.25
$$

---

## Step 5: Build the First Decision Tree

Using the **Gradients** and **Hessians**, XGBoost builds the first decision tree.

Example:

```text
           Age < 30 ?
          /          \
      Leaf 1       Leaf 2
```

The tree groups similar gradients together.

---

## Step 6: Calculate Similarity Score

Each leaf receives a Similarity Score.

$$
\text{Similarity}
=================

\frac{\left(\sum g_i\right)^2}
{\sum h_i+\lambda}
$$

where,

* $g_i$ = Gradient
* $h_i$ = Hessian
* $\lambda$ = L2 Regularization

A higher Similarity Score indicates a better leaf.

---

## Step 7: Calculate Gain

Every possible split is evaluated using the Gain formula.

$$
\text{Gain}
===========

\frac{1}{2}
\left(
\text{Left}
+
\text{Right}
------------

\text{Parent}
\right)
-\gamma
$$

where,

* Left = Left Similarity Score
* Right = Right Similarity Score
* Parent = Parent Similarity Score
* $\gamma$ = Minimum loss reduction

The split with the **highest Gain** is selected.

---

## Step 8: Calculate Leaf Weight

Each leaf predicts an output value called the **Leaf Weight**.

$$
w
=

-\frac{\sum g_i}
{\sum h_i+\lambda}
$$

This value is used to update the prediction.

---

## Step 9: Update the Prediction

Update the Log-Odds using the Learning Rate.

$$
\text{New Prediction}
=====================

\text{Old Prediction}
+
\eta \times w
$$

where,

* $\eta$ = Learning Rate
* $w$ = Leaf Weight

Example:

Old Prediction = 0

Leaf Weight = 2

Learning Rate = 0.3

$$
0+0.3\times2=0.6
$$

---

## Step 10: Convert Updated Log-Odds to Probability

Use the Sigmoid Function again.

$$
P=\frac{1}{1+e^{-0.6}}
$$

$$
P\approx0.645
$$

The prediction improves from

$$
0.50 \rightarrow 0.645
$$

---

## Step 11: Repeat the Process

The following steps are repeated until the model stops improving or the maximum number of trees is reached:

1. Calculate Gradient
2. Calculate Hessian
3. Build a Decision Tree
4. Calculate Similarity Score
5. Calculate Gain
6. Compute Leaf Weight
7. Update Prediction
8. Repeat

---

## Complete Workflow

```text
Training Data
      │
      ▼
Initial Prediction (0.5)
      │
      ▼
Convert to Log-Odds
      │
      ▼
Calculate Gradient
      │
      ▼
Calculate Hessian
      │
      ▼
Build Decision Tree
      │
      ▼
Calculate Similarity Score
      │
      ▼
Calculate Gain
      │
      ▼
Compute Leaf Weight
      │
      ▼
Update Prediction
      │
      ▼
Convert to Probability
      │
      ▼
Repeat Until All Trees Are Built
      │
      ▼
Final Prediction
```
