# When to Use Simple Linear Regression

Simple Linear Regression is used when you want to model the relationship between **one independent variable (X)** and **one dependent variable (Y)**, assuming the relationship is approximately linear.

## You should use Simple Linear Regression when:

### 1. There is only one independent variable
- You have a single predictor (input feature) and one target variable.
- Example:
  - Hours Studied → Exam Score
  - House Size → House Price

### 2. The relationship is approximately linear
- As the independent variable increases or decreases, the dependent variable changes in a roughly straight-line pattern.
- A scatter plot should show points clustered around a straight line.

### 3. You want to predict a continuous value
- The target variable should be numerical (continuous), not categorical.
- Examples:
  - Salary
  - Temperature
  - Sales
  - Price

### 4. You want to understand the impact of one variable on another
- It helps answer questions like:
  - How much does the exam score increase for every additional hour studied?
  - How much does house price increase for every additional square foot?

### 5. The observations are independent
- Each data point should be independent of the others.

---

## Real-World Examples

- Predicting exam score from hours studied.
- Predicting house price from house size.
- Predicting fuel consumption from distance traveled.
- Predicting electricity consumption from temperature.
- Predicting employee salary from years of experience.

---

## When NOT to Use Simple Linear Regression

❌ **Multiple factors affect the outcome**
- Example:
  - House Price depends on:
    - Size
    - Location
    - Number of bedrooms
    - Age of the house
- Use **Multiple Linear Regression** instead.

❌ **The relationship is non-linear**
- If the data forms a curve rather than a straight line, Simple Linear Regression is not appropriate.
- Consider Polynomial Regression or another non-linear model.

❌ **The target variable is categorical**
- Example:
  - Spam or Not Spam
  - Pass or Fail
- Use classification algorithms such as Logistic Regression instead.

---

## Summary

| Condition | Simple Linear Regression? |
|-----------|---------------------------|
| One independent variable | ✅ Yes |
| Continuous target variable | ✅ Yes |
| Linear relationship | ✅ Yes |
| Multiple input variables | ❌ No (Use Multiple Linear Regression) |
| Curved relationship | ❌ No (Use Polynomial/Non-linear Regression) |
| Categorical target | ❌ No (Use Classification) |

### Rule of Thumb

> **Use Simple Linear Regression when you have one input variable, one continuous output variable, and their relationship is approximately linear.**
