# Logistic Regression Interview Questions and Answers

## 1. What is Logistic Regression, and how does it differ from Linear Regression?

**Answer:**

Logistic Regression is a supervised machine learning algorithm used for **classification** problems. It predicts the probability that an input belongs to a particular class (0 or 1).

### Difference between Linear Regression and Logistic Regression

| Linear Regression           | Logistic Regression       |
| --------------------------- | ------------------------- |
| Used for Regression         | Used for Classification   |
| Predicts continuous values  | Predicts class labels     |
| Output ranges from (-∞, +∞) | Output ranges from (0, 1) |
| Uses a linear equation      | Uses the Sigmoid function |

---

## 2. Why can't we use Linear Regression for Classification Problems?

**Answer:**

Linear Regression is not suitable because:

* It predicts values outside the range of 0 and 1.
* Classification requires probabilities between 0 and 1.
* It is sensitive to outliers.
* It assumes a linear relationship, which is not suitable for classification.

Therefore, Logistic Regression is preferred.

---

## 3. Explain the Sigmoid (Logistic) Function and why it is used.

**Answer:**

The Sigmoid function converts any real number into a value between **0 and 1**, making it suitable for probability prediction.

### Formula

[
\sigma(z)=\frac{1}{1+e^{-z}}
]

where

[
z=\beta_0+\beta_1x_1+\beta_2x_2+\cdots+\beta_nx_n
]

### Why is it used?

* Converts output into probability.
* Helps classify data into two classes.
* Smooth and differentiable.

---

## 4. What is the Cost Function used in Logistic Regression? Why not use MSE?

**Answer:**

Logistic Regression uses **Log Loss (Binary Cross-Entropy)**.

### Formula

[
Loss=-(y\log(\hat y)+(1-y)\log(1-\hat y))
]

### Why not MSE?

* MSE creates a non-convex cost function.
* Gradient Descent may not converge efficiently.
* Log Loss is convex and easier to optimize.

---

## 5. Derive the Cost Function (Log Loss).

**Answer:**

Logistic Regression uses **Maximum Likelihood Estimation (MLE)**.

The likelihood function is converted into the **Negative Log-Likelihood**, giving the Log Loss:

[
J(\theta)= -\frac{1}{m}\sum_{i=1}^{m}
\left[y_i\log(\hat y_i)+(1-y_i)\log(1-\hat y_i)\right]
]

The objective is to **minimize this cost**.

---

## 6. How are the coefficients estimated?

**Answer:**

The coefficients are estimated using:

* Maximum Likelihood Estimation (MLE)
* Gradient Descent
* Newton-Raphson Method (used in many libraries)

The goal is to maximize the likelihood of observing the data.

---

## 7. What is the interpretation of coefficients?

**Answer:**

Each coefficient represents the **change in log-odds** of the target variable for a one-unit increase in the predictor.

After exponentiation:

[
Odds\ Ratio=e^{\beta}
]

* Odds Ratio > 1 → increases probability
* Odds Ratio < 1 → decreases probability

---

## 8. What are Odds and Log-Odds?

**Answer:**

### Odds

[
Odds=\frac{P}{1-P}
]

### Log-Odds

[
Logit=\log\left(\frac{P}{1-P}\right)
]

Logistic Regression models the **log-odds** instead of probability directly.

---

## 9. What are the assumptions of Logistic Regression?

**Answer:**

* Binary target variable
* Independent observations
* No multicollinearity
* Linear relationship between features and log-odds
* Large sample size
* No extreme outliers

---

## 10. How do you handle Multicollinearity?

**Answer:**

Methods include:

* Calculate VIF (Variance Inflation Factor)
* Remove highly correlated features
* Combine correlated variables
* Use L1 or L2 Regularization

---

## 11. How do you evaluate a Logistic Regression model?

**Answer:**

Common evaluation metrics:

* Accuracy
* Precision
* Recall
* F1-Score
* Confusion Matrix
* ROC-AUC Score

**Note:** For imbalanced datasets, Accuracy alone is not enough. Use Precision, Recall, F1-Score, and ROC-AUC.

---

## 12. How do you handle Class Imbalance?

**Answer:**

Methods include:

* Oversampling
* Undersampling
* SMOTE
* Class Weight adjustment
* Threshold tuning
* Use Precision, Recall, F1-score, and ROC-AUC instead of Accuracy.

---

## 13. What is Regularization? Explain L1 vs L2.

**Answer:**

Regularization helps prevent overfitting.

### L1 Regularization (Lasso)

* Adds absolute values of coefficients.
* Performs feature selection.
* Some coefficients become exactly zero.

### L2 Regularization (Ridge)

* Adds squared values of coefficients.
* Shrinks coefficients.
* Does not eliminate features.

---

## 14. How do you choose the optimal threshold instead of 0.5?

**Answer:**

Methods:

* ROC Curve
* Precision-Recall Curve
* Youden's Index
* Business requirements
* Cost of False Positives and False Negatives

The threshold depends on the problem rather than always using 0.5.

---

## 15. Can Logistic Regression be used for Multi-Class Classification?

**Answer:**

Yes.

Two common approaches:

### One-vs-Rest (OvR)

* Creates one classifier for each class.
* Most common method.

### Multinomial Logistic Regression (Softmax)

* Predicts all classes simultaneously.
* Used when classes are mutually exclusive.

---

# Quick Interview Tips

* Logistic Regression is a **classification algorithm**.
* Uses the **Sigmoid Function**.
* Optimizes using **Maximum Likelihood Estimation (MLE)**.
* Uses **Log Loss (Binary Cross-Entropy)**.
* Coefficients represent **log-odds**.
* (e^\beta) gives the **Odds Ratio**.
* Evaluate using **Accuracy, Precision, Recall, F1-Score, ROC-AUC, and Confusion Matrix**.
* Use **L1/L2 Regularization** to reduce overfitting.
* For multiclass classification, use **One-vs-Rest (OvR)** or **Multinomial Logistic Regression**.
