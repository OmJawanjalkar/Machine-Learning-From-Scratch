# 🌳 Gradient Boosting Regressor Using 3 Weak Learners

Gradient Boosting Regressor builds multiple **weak decision trees sequentially**, where each new tree learns from the **errors (residuals)** made by the previous trees. By gradually correcting these mistakes, the model becomes more accurate after every iteration.

---

# 📊 Training Dataset

Suppose we want to predict the **price** of a house based on its **size**.

| Size | Price |
| ---- | ----- |
| 1    | 100   |
| 2    | 200   |
| 3    | 300   |

---

# Step 1: Train the First Weak Learner (Tree #1)

The first weak learner is a simple decision tree.

It predicts the **average price** of all training samples.

Average Price

```text
(100 + 200 + 300) / 3 = 200
```

Therefore, Tree #1 predicts **200** for every observation.

| Size | Actual Price | Prediction (Tree #1) | Residual (Actual − Prediction) |
| ---- | ------------ | -------------------- | ------------------------------ |
| 1    | 100          | 200                  | -100                           |
| 2    | 200          | 200                  | 0                              |
| 3    | 300          | 200                  | +100                           |

### Observation

* Size = 1 is overestimated by **100**.
* Size = 2 is predicted correctly.
* Size = 3 is underestimated by **100**.

These prediction errors are called **Residuals**.

---

# Step 2: Train the Second Weak Learner (Tree #2)

Instead of predicting the original prices, the second tree learns the residuals produced by Tree #1.

Tree #2 learns:

| Size | Residual |
| ---- | -------- |
| 1    | -100     |
| 2    | 0        |
| 3    | +100     |

Now we add these corrections to the first tree.

### Updated Predictions

| Size | Tree #1 | Tree #2 | Final Prediction |
| ---- | ------- | ------- | ---------------- |
| 1    | 200     | -100    | 100 ✅            |
| 2    | 200     | 0       | 200 ✅            |
| 3    | 200     | +100    | 300 ✅            |

After adding the second tree, every prediction becomes correct.

---

# Step 3: Train the Third Weak Learner (Tree #3)

Since the predictions are already perfect, there are no residual errors remaining.

| Size | Residual |
| ---- | -------- |
| 1    | 0        |
| 2    | 0        |
| 3    | 0        |

Therefore, the third tree has nothing new to learn and contributes almost nothing to the final model.

---

# ✅ Final Gradient Boosting Model

The final prediction is obtained by adding the outputs of all weak learners.

```text
Final Prediction
      =
Tree #1
   +
Tree #2
   +
Tree #3
```

Final Predictions:

| Size | Predicted Price |
| ---- | --------------- |
| 1    | 100             |
| 2    | 200             |
| 3    | 300             |

The predictions exactly match the actual prices.

---

# 🌱 Important Terms in Gradient Boosting Regressor

## 1. Weak Learner

A weak learner is a simple model, usually a **small Decision Tree**.

* It performs only slightly better than random guessing.
* Many weak learners together form a powerful prediction model.

---

## 2. Ensemble

An **ensemble** is a collection of multiple weak learners.

Instead of relying on a single tree, Gradient Boosting combines many trees sequentially to improve prediction accuracy.

---

## 3. Loss Function

A **Loss Function** measures how far the predictions are from the actual values.

Common loss functions used in regression include:

* **Mean Squared Error (MSE)** – Gives more penalty to large errors.
* **Mean Absolute Error (MAE)** – Measures the average absolute difference between predicted and actual values.

Gradient Boosting tries to minimize this loss after every iteration.

---

## 4. Residuals (Errors)

Residuals represent the prediction errors.

Residual Formula:

```text
Residual = Actual Value − Predicted Value
```

Each new decision tree learns these residuals instead of learning the original target values.

---

## 5. Gradient

The **Gradient** indicates the direction in which the loss decreases most rapidly.

Each new tree learns this direction so that the model gradually reduces prediction errors.

---

## 6. Learning Rate

The **Learning Rate** controls how much each new tree contributes to the final prediction.

* Small learning rate → Slower learning but better generalization.
* Large learning rate → Faster learning but may overfit.

Common values:

* 0.01
* 0.05
* 0.10
* 0.20

---

## 7. Number of Estimators

The **Number of Estimators** represents the total number of weak learners (decision trees).

* More trees usually improve performance.
* Too many trees may increase computation time and cause overfitting.

Typical values:

* 50
* 100
* 200
* 300

---

## 8. Tree Depth

Tree depth determines the complexity of each weak learner.

* Small depth → Better generalization.
* Large depth → More complex model with higher overfitting risk.

Typical values:

* 1
* 2
* 3
* 5

---

## 9. Subsampling

Subsampling means using only a random portion of the training data to build each tree.

Benefits include:

* Reduces overfitting
* Improves generalization
* Makes the model more robust

Typical values:

* 0.6
* 0.8
* 1.0

---

## 10. Regularization

Regularization includes techniques that prevent the model from overfitting.

Common methods include:

* Limiting tree depth
* Using a smaller learning rate
* Using fewer features per tree
* Applying subsampling
* Reducing the number of estimators

These techniques help the model perform better on unseen data.

---

## 11. Additive Model

Gradient Boosting is called an **Additive Model** because it builds the final prediction by **adding one weak learner at a time**.

Mathematically,

```text
Final Model
=
Tree₁
+
Tree₂
+
Tree₃
+
...
+
Treeₙ
```

Every new tree corrects the mistakes of the previous trees, gradually improving prediction accuracy.

---

# ✅ Key Takeaways

* Gradient Boosting Regressor builds decision trees **sequentially**.
* Each new tree learns the **residual errors** of the previous trees.
* The model minimizes a **loss function** such as Mean Squared Error (MSE).
* Weak learners are usually shallow decision trees.
* The final prediction is the **sum of all weak learners**.
* Proper tuning of learning rate, tree depth, number of estimators, and subsampling helps improve performance while reducing overfitting.
