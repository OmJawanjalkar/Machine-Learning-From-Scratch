# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

## What is DBSCAN?

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** is an **unsupervised machine learning clustering algorithm** that groups together data points that are **closely packed (high density)** and marks points that are far away from others as **outliers (noise)**.

Unlike **K-Means**, DBSCAN **does not require you to specify the number of clusters beforehand**.

---

# Simple Definition

Imagine people standing in a park.

- If many people stand close together, they form a **group (cluster)**.
- If one person stands alone far away, they are considered an **outlier (noise)**.

DBSCAN works exactly the same way.

---

# Real-Life Example

Imagine you are looking at restaurants in a city.

- 🍽️ Many restaurants in the city center → **Cluster**
- 🛍️ Another group inside a shopping mall → **Cluster**
- 🛣️ One isolated restaurant on a highway → **Noise (Outlier)**

DBSCAN automatically discovers these groups.

---

# Main Idea

DBSCAN forms clusters based on **density**.

Instead of calculating the center of a cluster like **K-Means**, DBSCAN asks:

> **"Are there enough neighboring points within a certain distance?"**

If **Yes**
- Grow the cluster.

If **No**
- Mark the point as **Noise**.

---

# Two Important Parameters

## 1. Epsilon ($\varepsilon$)

**Epsilon ($\varepsilon$)** defines the radius around each point.

Think of drawing a circle around every data point.

```
      ε Radius

         ○
      ○  ●  ○
         ○
```

Every point inside the circle is considered a **neighbor**.

---

## 2. MinPts (Minimum Points)

**MinPts** is the minimum number of neighboring points required inside the $\varepsilon$-radius to form a dense region.

### Example

**MinPts = 5**

```
      ○
   ○  ●  ○
      ○
```

There are **5 points**, so this becomes a **Cluster**.

If only two points exist:

```
●      ○
```

There is not enough density.

→ This point becomes **Noise**.

---

# Types of Points in DBSCAN

There are **three types of points**.

---

## 1. Core Point

A point having at least **MinPts** neighbors inside the $\varepsilon$-radius.

```
       ○
   ○   ●   ○
       ○
```

Enough nearby points.

→ **Core Point**

---

## 2. Border Point

A point that does **not** have enough neighbors itself but is connected to a **Core Point**.

```
      ●
   ○  ○  ○
          ○
           B
```

Border points belong to the cluster but cannot expand it.

---

## 3. Noise Point (Outlier)

A point that does not belong to any dense region.

```
Cluster

○ ○ ○ ○

          X
```

**X** is considered **Noise**.

# Mathematical Idea

For a point $p$, the $\varepsilon$-neighborhood is defined as

$$
N_{\varepsilon}(p)
=
\{q \mid \operatorname{dist}(p,q) \le \varepsilon\}
$$

where

- $\operatorname{dist}(p,q)$ = Distance between two points
- $\varepsilon$ = Neighborhood radius

A point is called a **Core Point** if

$$
|N_{\varepsilon}(p)| \ge \text{MinPts}
$$

where

- $|N_{\varepsilon}(p)|$ = Number of neighboring points
- **MinPts** = Minimum number of required neighbors

---

# DBSCAN Algorithm

For each point in the dataset:

```
Start

│

├── Is the point visited?
│
├── No
│
├── Mark it as Visited
│
├── Find all neighbors within ε
│
├── Are Neighbors ≥ MinPts?
│      │
│      ├── Yes
│      │     ├── Create a New Cluster
│      │     └── Expand the Cluster
│      │
│      └── No
│            └── Mark as Noise
│
└── Continue until all points are visited
```

---

# Advantages

- ✅ No need to specify the number of clusters beforehand.
- ✅ Automatically detects outliers.
- ✅ Can discover clusters of arbitrary (non-spherical) shapes.
- ✅ Works well when clusters are clearly separated.

---

# Disadvantages

- ❌ Choosing appropriate values of $\varepsilon$ and **MinPts** can be difficult.
- ❌ Performance decreases for high-dimensional data.
- ❌ Struggles when clusters have significantly different densities.

---

# DBSCAN vs K-Means

| Feature | DBSCAN | K-Means |
|---------|---------|---------|
| Need to specify number of clusters | ❌ No | ✅ Yes (K) |
| Detects Outliers | ✅ Yes | ❌ No |
| Cluster Shape | Any Shape | Mostly Spherical |
| Uses Centroid | ❌ No | ✅ Yes |
| Handles Noise | ✅ Excellent | ❌ Poor |
| Sensitive to Initialization | ❌ No | ✅ Yes |
| Works with Different Densities | ❌ Not Always | ❌ Generally No |

---

# Time Complexity

Without a spatial index:

$$
O(n^2)
$$

With efficient indexing (KD-Tree or Ball Tree):

$$
O(n \log n)
$$

where

- $n$ = Number of data points

---

# When Should You Use DBSCAN?

Use DBSCAN when:

- ✅ The number of clusters is unknown.
- ✅ Your dataset contains outliers.
- ✅ Clusters have irregular or complex shapes.
- ✅ Automatic noise detection is important.

---

# When Should You Avoid DBSCAN?

Avoid DBSCAN when:

- ❌ Clusters have very different densities.
- ❌ The dataset is high-dimensional without dimensionality reduction.
- ❌ Selecting an appropriate $\varepsilon$ value is difficult.

---

# Key Takeaways

- **DBSCAN** is a **density-based clustering algorithm**.
- It automatically discovers the **number of clusters**.
- It classifies points into **Core**, **Border**, and **Noise**.
- It performs well on **arbitrary-shaped clusters**.
- It is one of the best clustering algorithms when **outlier detection** is important.
