# Little Math

## 1. Bayes' Theorem

Bayes' Theorem is used to find the probability of an event when we already know some related information.

### Formula

$$
P(A|B)=\frac{P(B|A)\times P(A)}{P(B)}
$$

Where:

- **P(A|B)** = Probability of A after B has happened.
- **P(B|A)** = Probability of B if A is true.
- **P(A)** = Probability of A.
- **P(B)** = Probability of B.

---

## 2. Simple Example

Suppose:

- 1% of emails are spam.
- 90% of spam emails contain the word **"offer"**.
- 20% of normal emails also contain the word **"offer"**.

If an email contains the word **"offer"**, Bayes' Theorem helps us find the probability that the email is actually spam.

---

## 3. Multiple Features

In machine learning, we usually have many features instead of one.

For example:

- offer
- win
- free
- money

Bayes' Theorem becomes:

$$
P(Class|x_1,x_2,...,x_n)
=
\frac{P(x_1,x_2,...,x_n|Class)\times P(Class)}
{P(x_1,x_2,...,x_n)}
$$

Where:

- **Class** = Output category (Spam or Not Spam)
- **x₁, x₂, ...** = Features

---

## 4. Why is it called "Naive"?

Finding the probability of many features together is difficult.

So Naive Bayes makes one simple assumption:

> **All features are independent of each other given the class.**

Because of this assumption,

$$
P(x_1,x_2,...,x_n|Class)
\approx
P(x_1|Class)\times P(x_2|Class)\times ... \times P(x_n|Class)
$$

This makes the calculations much easier.

---

## 5. Naive Bayes Formula

The final formula becomes:

$$
P(Class|x_1,x_2,...,x_n)
\propto
P(Class)\times\prod_{i=1}^{n}P(x_i|Class)
$$

Here:

- **P(Class)** = Prior Probability
- **P(xᵢ|Class)** = Likelihood
- **∏** = Multiply the probabilities of all features.

The denominator is the same for every class, so we ignore it.

### Prediction Steps

1. Calculate the probability for each class.
2. Multiply the prior probability with the likelihood of each feature.
3. Choose the class with the highest probability.

---

# About the Algorithm

## What is Naive Bayes?

Naive Bayes is a **probability-based classification algorithm**.

It predicts which category a data point belongs to by calculating probabilities.

It is called:

- **Naive** because it assumes all features are independent.
- **Bayes** because it uses Bayes' Theorem.

### How it Works

1. Take the input features.
2. Calculate the probability for each class.
3. Select the class with the highest probability.

**In short:**  
Naive Bayes predicts the most likely class using probability.

---

# Types of Naive Bayes

## 1. Bernoulli Naive Bayes

### Used For

Binary (Yes/No or 0/1) features.

### Example

Email contains the word **"offer"**?

- Yes = 1
- No = 0

It only checks whether the word is present or not.

---

## 2. Multinomial Naive Bayes

### Used For

Count data.

Mostly used in text classification.

### Example

Word frequencies:

- offer = 3
- win = 2
- free = 1

It uses the number of times each word appears.

---

## 3. Gaussian Naive Bayes

### Used For

Continuous numerical data.

It assumes the data follows a **Normal (Gaussian) Distribution**.

### Example

Predict a fruit using:

- Weight
- Diameter

These values are continuous numbers.

---

# Quick Summary

| Type | Data Type | Example |
|-------|-----------|---------|
| Bernoulli NB | Binary (0/1) | Word present or absent |
| Multinomial NB | Counts | Word frequency in text |
| Gaussian NB | Continuous values | Height, Weight, Temperature |

---

## Advantages

- Easy to understand.
- Very fast.
- Works well with text data.
- Performs well even with small datasets.

---

## Disadvantages

- Assumes features are independent.
- Performance may decrease if features are highly related.
- Gaussian Naive Bayes assumes data follows a normal distribution.
