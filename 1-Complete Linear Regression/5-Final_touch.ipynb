# Regression Models and Evaluation Metrics: A Quick Overview

This section provides a high-level overview of the most common regression algorithms, model complexity concepts, evaluation metrics, and goodness-of-fit measures used in Machine Learning.

---

# Regression Models

Regression algorithms are supervised learning techniques used to **predict continuous numerical values** by learning the relationship between one or more independent variables (features) and a dependent variable (target).

## 1. Simple Linear Regression

Simple Linear Regression models the relationship between **one independent variable** and **one dependent variable** using a straight-line equation.

### Characteristics

* One input feature
* Continuous target variable
* Linear relationship between input and output

**Example**

Predict a student's exam score using the number of hours studied.

---

## 2. Multiple Linear Regression

Multiple Linear Regression extends Simple Linear Regression by incorporating **two or more independent variables** to predict a single continuous target.

### Characteristics

* Multiple input features
* Continuous target variable
* Linear relationship between each feature and the target

**Example**

Predict a student's exam score using:

* Hours Studied
* Attendance
* Sleep Duration

---

## 3. Polynomial Regression

Polynomial Regression is used when the relationship between the features and the target is **non-linear**. It introduces polynomial terms (e.g., $x^2$, $x^3$) while still using a linear regression model on the transformed features.

### Characteristics

* Models curved relationships
* Captures higher-order feature interactions
* More flexible than Linear Regression

**Example**

Predict plant growth over time, where growth accelerates initially and gradually slows down.

---

# Model Complexity

A machine learning model should learn the underlying patterns in the data without memorizing noise.

## 1. Underfitting

Underfitting occurs when the model is **too simple** to capture the underlying relationship.

### Characteristics

* High training error
* High testing error
* Poor predictive performance

**Example**

Using a straight line to model a highly curved dataset.

---

## 2. Overfitting

Overfitting occurs when the model becomes **too complex** and starts learning random noise in the training data.

### Characteristics

* Very low training error
* High testing error
* Poor generalization

**Example**

Using a very high-degree polynomial that passes through nearly every training point.

---

### Goal

A good machine learning model should achieve an appropriate balance between **underfitting** and **overfitting**, allowing it to generalize well to unseen data.

---

# Regression Evaluation Metrics

Evaluation metrics measure how accurately the model predicts the target variable.

## 1. Mean Absolute Error (MAE)

Measures the average absolute difference between actual and predicted values.

### Characteristics

* Easy to interpret
* Less sensitive to outliers
* Errors are measured in the original unit

**Interpretation**

> On average, predictions differ from the actual values by **X units**.

---

## 2. Mean Squared Error (MSE)

Measures the average squared difference between actual and predicted values.

### Characteristics

* Penalizes larger errors more heavily
* Sensitive to outliers
* Commonly used during model optimization

---

## 3. Root Mean Squared Error (RMSE)

RMSE is the square root of the Mean Squared Error.

### Characteristics

* Same unit as the target variable
* Easier to interpret than MSE
* Widely used for regression model evaluation

---

# Goodness-of-Fit Metrics

These metrics evaluate how well the regression model explains the variation in the target variable.

## 1. R² Score (Coefficient of Determination)

R² measures the proportion of variance in the target variable explained by the regression model.

### Interpretation

* **R² = 1** → Perfect prediction
* **R² = 0** → Model performs no better than predicting the mean
* **R² < 0** → Model performs worse than simply predicting the mean

---

## 2. Adjusted R² Score

Adjusted R² modifies the R² score by penalizing unnecessary independent variables.

### Why is it useful?

In Multiple Linear Regression, adding more features always increases or maintains the R² score, even if those features provide no meaningful information.

Adjusted R² increases **only when newly added features genuinely improve the model**, making it a more reliable metric for comparing models with different numbers of predictors.

---

# Key Takeaways

* **Simple Linear Regression** is used for a single predictor.
* **Multiple Linear Regression** handles multiple predictors simultaneously.
* **Polynomial Regression** models non-linear relationships through polynomial feature expansion.
* **Underfitting** occurs when the model is too simple, whereas **Overfitting** occurs when the model becomes overly complex.
* **MAE**, **MSE**, and **RMSE** quantify prediction errors from different perspectives.
* **R²** and **Adjusted R²** evaluate how effectively the model explains the variability in the target variable.

> **Rule of Thumb:** Build the simplest model that accurately captures the underlying data patterns while maintaining strong generalization performance on unseen data.
