## When to Use Naive Bayes

Naive Bayes is a **supervised machine learning algorithm** used for **classification** tasks. It is simple, fast, and works well when the features are independent.

### Use Naive Bayes When:
- The problem is a **classification** task.
- Working with **text data** (Spam Detection, Sentiment Analysis).
- The dataset is **small or medium-sized**.
- Fast training and prediction are required.
- Features are approximately **independent**.


---




### When Not to Use Naive Bayes:
- For **regression** problems.
- When features are **highly correlated**.
- When the data has **complex relationships** between features.




> **Key Point:** Naive Bayes is a fast and efficient classification algorithm, especially for text classification and high-dimensional datasets.
