# When to Use KNN Classifier and KNN Regressor

## KNN Classifier

### ✅ When to Use

* When the target variable is **categorical** (e.g., Yes/No, Spam/Not Spam, Cat/Dog).
* When similar data points are expected to belong to the same class.
* When working with **small to medium-sized datasets**.
* When the features are properly **scaled**.

### ❌ When Not to Use

* When the dataset is very large, as prediction becomes slow.
* When the data has many irrelevant features.
* When the dataset has high dimensionality (Curse of Dimensionality).
* When the classes are highly imbalanced.
* When the features are not scaled.

---

## KNN Regressor

### ✅ When to Use

* When the target variable is **continuous (numeric)** (e.g., House Price, Salary, Temperature).
* When nearby data points have similar numerical values.
* When the relationship between features and the target is non-linear.
* When working with **small to medium-sized datasets**.
* When the features are properly **scaled**.

### ❌ When Not to Use

* When the dataset is very large.
* When the target variable contains many outliers.
* When the data has high dimensionality.
* When the features are not scaled.
* When fast prediction is required for real-time applications.

---

## Quick Rule

| Target Variable             | Algorithm to Use   |
| --------------------------- | ------------------ |
| Categorical (Class Labels)  | **KNN Classifier** |
| Continuous (Numeric Values) | **KNN Regressor**  |

> **Remember:**
>
> * **Classification → KNN Classifier**
> * **Regression → KNN Regressor**
