# Top 15 Linear Regression Interview Questions

These are the most frequently asked interview questions on **Linear Regression** for **Data Scientist**, **Machine Learning Engineer**, and **Data Analyst** roles.

---

## 1. What is Linear Regression?

* What is Linear Regression?
* Why is it called a linear model?
* What type of problem does it solve?
* Give a real-world example of Linear Regression.

---

## 2. What are the assumptions of Linear Regression?

Explain the following assumptions:

* Linearity
* Independence of observations
* Homoscedasticity
* Normality of residuals
* No Multicollinearity
* No Auto-correlation (when applicable)

---

## 3. Explain the Cost Function in Linear Regression.

* What is a cost function?
* Why do we need a cost function?
* What is Mean Squared Error (MSE)?
* Why are errors squared?
* Why is MSE preferred over MAE for optimization?

---

## 4. What is Ordinary Least Squares (OLS)?

* What is OLS?
* How does OLS estimate regression coefficients?
* Why is it called "Least Squares"?
* Why don't we minimize the absolute error?

---

## 5. What is the difference between Simple Linear Regression and Multiple Linear Regression?

Explain the differences in terms of:

* Number of independent variables
* Mathematical equation
* Use cases
* Examples

---

## 6. What is Polynomial Regression?

* What is Polynomial Regression?
* Why do we use Polynomial Regression?
* Is Polynomial Regression a linear model?
* How is it different from Linear Regression?

---

## 7. Explain Underfitting and Overfitting.

* What is Underfitting?
* What is Overfitting?
* Causes
* Effects
* How can they be prevented?

---

## 8. Explain MAE, MSE, and RMSE.

* What is Mean Absolute Error (MAE)?
* What is Mean Squared Error (MSE)?
* What is Root Mean Squared Error (RMSE)?
* Which metric is more sensitive to outliers?
* When should each metric be used?

---

## 9. What is the R² Score?

* What is R²?
* What does an R² value of 0.85 indicate?
* Can R² be negative?
* What are the limitations of R²?

---

## 10. What is Adjusted R²?

* Why do we need Adjusted R²?
* Difference between R² and Adjusted R².
* Why doesn't Adjusted R² always increase?
* When should we use Adjusted R²?

---

## 11. What is Multicollinearity?

* What is Multicollinearity?
* Why is it a problem?
* How can it be detected?
* What is Variance Inflation Factor (VIF)?
* How can Multicollinearity be reduced?

---

## 12. Explain Ridge, Lasso, and Elastic Net Regression.

Explain:

* Ridge Regression (L2 Regularization)
* Lasso Regression (L1 Regularization)
* Elastic Net Regression (L1 + L2)

Also explain:

* Differences
* Advantages
* Disadvantages
* Use cases

---

## 13. What is Cross-Validation?

* Why is Cross-Validation required?
* What are the different types of Cross-Validation?
* What is K-Fold Cross-Validation?
* Why is K-Fold commonly preferred?
* Difference between Train-Test Split and Cross-Validation.

---

## 14. What is Hyperparameter Tuning?

Explain:

* Hyperparameters vs Parameters
* Why Hyperparameter Tuning is required
* Grid Search
* Random Search
* RidgeCV
* LassoCV
* ElasticNetCV

---

## 15. Explain the complete workflow of a Linear Regression project.

A complete workflow should include:

1. Business Problem Understanding
2. Data Collection
3. Exploratory Data Analysis (EDA)
4. Data Cleaning
5. Handling Missing Values
6. Outlier Detection
7. Feature Engineering
8. Feature Selection
9. Data Splitting
10. Model Training
11. Model Evaluation
12. Cross-Validation
13. Hyperparameter Tuning
14. Model Selection
15. Model Deployment

---

# Bonus Interview Questions

These are also commonly asked during interviews.

1. Why do we square the errors instead of taking the absolute value?

2. Why is Linear Regression called a Parametric Model?

3. Can Linear Regression be used for Classification?

4. What are Residuals?

5. What is Heteroscedasticity?

6. What is Homoscedasticity?

7. What is the difference between Parameters and Hyperparameters?

8. Why is Feature Scaling important for Ridge and Lasso Regression?

9. What is the Bias-Variance Tradeoff?

10. How do you evaluate whether a Linear Regression model is performing well?

11. How do you detect Outliers in Linear Regression?

12. What happens if the assumptions of Linear Regression are violated?

13. Why is Cross-Validation preferred over a single Train-Test Split?

14. What is the difference between RidgeCV, LassoCV, and ElasticNetCV?

15. When would you choose Ridge, Lasso, or Elastic Net Regression?

---

# Interview Tip

If you can confidently explain these **15 core questions** and **15 bonus questions** with their concepts, mathematical intuition, practical examples, and implementation in Scikit-learn, you will be well prepared for **Data Analyst**, **Machine Learning Engineer**, and **Data Scientist** interviews focusing on Linear Regression.
