# 📌 When to Use Gradient Boosting Classifier

Use **Gradient Boosting Classifier** when:

- The target variable is **categorical** (e.g., Yes/No, Spam/Not Spam, Disease/No Disease).
- High prediction accuracy is more important than training speed.
- The dataset has complex, non-linear relationships.
- You want a powerful ensemble model that reduces bias and improves performance.

### Examples
- Customer Churn Prediction
- Fraud Detection
- Email Spam Classification
- Disease Diagnosis
- Loan Approval Prediction
- Sentiment Analysis

# 📌 When to Use Gradient Boosting Regressor

Use **Gradient Boosting Regressor** when:

- The target variable is **continuous** (numeric values).
- You need highly accurate predictions.
- The data contains complex patterns that linear models cannot capture.
- You want to minimize prediction errors using boosting.

### Examples
- House Price Prediction
- Sales Forecasting
- Stock Price Prediction
- Temperature Prediction
- Energy Consumption Forecasting
- Insurance Cost Prediction
