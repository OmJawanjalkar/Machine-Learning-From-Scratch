# Support Vector Machine (SVM) - Top 20 Interview Questions and Answers

## 1. What is Support Vector Machine (SVM)?

**Answer:**
Support Vector Machine (SVM) is a supervised machine learning algorithm used mainly for classification problems. It finds the optimal hyperplane that separates different classes while maximizing the margin between them. SVM can also be used for regression tasks using Support Vector Regression (SVR).

---

## 2. Why is SVM called a Maximum Margin Classifier?

**Answer:**
SVM is called a Maximum Margin Classifier because it selects the hyperplane that maximizes the distance (margin) between the two classes. A larger margin helps improve the model's ability to generalize to unseen data.

---

## 3. What is a Hyperplane?

**Answer:**
A hyperplane is the decision boundary that separates different classes. In two dimensions, it is a line; in three dimensions, it is a plane; and in higher dimensions, it is called a hyperplane.

---

## 4. What are Support Vectors?

**Answer:**
Support vectors are the data points closest to the decision boundary. They are the most important observations because they determine the position of the optimal hyperplane.

---

## 5. What is the Margin in SVM?

**Answer:**
The margin is the distance between the hyperplane and the nearest data points (support vectors). SVM aims to maximize this margin to improve prediction accuracy.

---

## 6. What is SVC?

**Answer:**
SVC (Support Vector Classifier) is the classification version of SVM. It is used to classify data into two or more classes.

---

## 7. What is SVR?

**Answer:**
SVR (Support Vector Regression) is the regression version of SVM. Instead of predicting classes, it predicts continuous numerical values while allowing a small acceptable error (epsilon).

---

## 8. What is the difference between SVC and SVR?

**Answer:**

| SVC                     | SVR                        |
| ----------------------- | -------------------------- |
| Used for Classification | Used for Regression        |
| Predicts Class Labels   | Predicts Continuous Values |
| Uses Decision Boundary  | Uses Best Fit Function     |

---

## 9. What is a Kernel in SVM?

**Answer:**
A kernel is a mathematical function that transforms data into a higher-dimensional space, making it easier to separate classes that are not linearly separable.

---

## 10. What is the Kernel Trick?

**Answer:**
The Kernel Trick allows SVM to solve non-linear classification problems without explicitly transforming the data into a higher-dimensional space, making computation more efficient.

---

## 11. What are the different types of Kernels?

**Answer:**
The four commonly used kernels are:

* Linear Kernel
* Polynomial Kernel
* RBF (Gaussian) Kernel
* Sigmoid Kernel

---

## 12. Which Kernel is used most frequently?

**Answer:**
The RBF (Radial Basis Function) Kernel is the most commonly used because it performs well on most non-linear datasets and usually provides better classification accuracy.

---

## 13. What is the C parameter in SVM?

**Answer:**
The C parameter controls the trade-off between maximizing the margin and minimizing classification errors.

* Large C → Smaller margin, fewer training errors, higher risk of overfitting.
* Small C → Larger margin, more training errors, better generalization.

---

## 14. What is Gamma (γ) in SVM?

**Answer:**
Gamma controls the influence of a single training example.

* High Gamma → Complex decision boundary, higher risk of overfitting.
* Low Gamma → Simpler decision boundary, lower risk of overfitting.

---

## 15. Why is Feature Scaling important in SVM?

**Answer:**
SVM calculates distances between data points. If features have different scales, larger values dominate the distance calculation. Therefore, feature scaling improves model performance and training stability.

---

## 16. What is the difference between SVM and Logistic Regression?

**Answer:**

| SVM                                   | Logistic Regression           |
| ------------------------------------- | ----------------------------- |
| Maximizes the margin                  | Estimates class probabilities |
| Uses support vectors                  | Uses all training samples     |
| Works well with high-dimensional data | Faster on very large datasets |
| Can use kernel functions              | Cannot directly use kernels   |

---

## 17. When should you use SVM?

**Answer:**
Use SVM when:

* The dataset is small to medium-sized.
* The data has many features.
* The classes are clearly separable.
* High classification accuracy is required.
* The data is non-linear (using kernel functions).

---

## 18. When should you NOT use SVM?

**Answer:**
Avoid using SVM when:

* The dataset is very large.
* The data contains a lot of noise or overlapping classes.
* Fast training is required.
* A highly interpretable model is needed.

---

## 19. What are the advantages of SVM?

**Answer:**

* High classification accuracy.
* Effective for high-dimensional data.
* Works well with small datasets.
* Can solve both linear and non-linear problems.
* Less prone to overfitting due to margin maximization.

---

## 20. What are the disadvantages of SVM?

**Answer:**

* Training is slow for very large datasets.
* Choosing the right kernel and hyperparameters can be difficult.
* Sensitive to noisy data.
* Requires feature scaling.
* Computationally expensive for large datasets.
