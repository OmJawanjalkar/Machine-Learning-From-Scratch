# Top 10 Naive Bayes Interview Questions

## 1. What is Naive Bayes?

**Answer:**
Naive Bayes is a **supervised machine learning classification algorithm** based on **Bayes' Theorem**. It predicts the most likely class by calculating probabilities and assumes that all features are independent of each other.

---

## 2. Why is it called "Naive" Bayes?

**Answer:**
It is called **Naive** because it assumes that all input features are **independent** of each other, even though this assumption is often not true in real-world data.

---

## 3. What is the formula of Naive Bayes?

**Answer:**

$$
P(Class|X)=\frac{P(X|Class)\times P(Class)}{P(X)}
$$

For prediction:

$$
P(Class|X)\propto P(Class)\times\prod_{i=1}^{n}P(x_i|Class)
$$

Where:

* **P(Class)** = Prior Probability
* **P(xᵢ|Class)** = Likelihood
* **P(Class|X)** = Posterior Probability

---

## 4. What are the different types of Naive Bayes?

**Answer:**

1. **Gaussian Naive Bayes** – Used for continuous numerical data.
2. **Multinomial Naive Bayes** – Used for count data, such as word frequencies.
3. **Bernoulli Naive Bayes** – Used for binary (0/1) features.

---

---

## 6. What are the advantages of Naive Bayes?

**Answer:**

* Simple and easy to implement.
* Very fast for training and prediction.
* Works well with high-dimensional datasets.
* Performs well on text classification problems.
* Requires less training data.

---

## 7. What are the disadvantages of Naive Bayes?

**Answer:**

* Assumes all features are independent.
* Accuracy may decrease when features are highly correlated.
* Suffers from the Zero Probability Problem.
* Gaussian Naive Bayes assumes a normal distribution for continuous features.

---

## 8. What is the Zero Probability Problem?

**Answer:**
If a feature never appears in the training data for a particular class, its probability becomes **0**. Since Naive Bayes multiplies probabilities, the final probability also becomes **0**.

**Solution:** Use **Laplace Smoothing (Add-One Smoothing)** to avoid zero probabilities.
