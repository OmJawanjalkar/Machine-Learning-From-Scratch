# When to Use Multiple Linear Regression

Multiple Linear Regression (MLR) is a supervised machine learning algorithm used to model the relationship between **one dependent variable (Y)** and **two or more independent variables (X₁, X₂, ..., Xₙ)**. It is appropriate when multiple factors collectively influence a continuous outcome.

## You should use Multiple Linear Regression when:

### 1. There are multiple independent variables

* The prediction depends on two or more input features.
* Examples:

  * House Size
  * Number of Bedrooms
  * House Age
  * Location Score

### 2. The target variable is continuous

* The dependent variable should represent a numerical value.
* Examples:

  * House Price
  * Salary
  * Sales Revenue
  * Temperature
  * Energy Consumption

### 3. You want to understand the combined effect of multiple factors

* Multiple Linear Regression estimates how each independent variable influences the target variable while keeping all other variables constant.
* This helps quantify the contribution of each feature to the prediction.

### 4. The relationship is approximately linear

* The relationship between each independent variable and the dependent variable should be approximately linear, assuming the other variables remain constant.

### 5. The observations are independent

* Each observation in the dataset should be independent of the others to ensure reliable model estimation.

## When NOT to Use Multiple Linear Regression

❌ **The relationship is non-linear**

* If the relationship between the predictors and the target is highly curved or complex, Multiple Linear Regression may not produce accurate predictions.
* Consider Polynomial Regression or other non-linear models.

❌ **High multicollinearity exists**

* If two or more independent variables are highly correlated, the model coefficients become unstable and difficult to interpret.

❌ **The target variable is categorical**

* Multiple Linear Regression is designed for continuous target variables.
* For categorical outcomes (e.g., Yes/No, Spam/Not Spam), use classification algorithms such as Logistic Regression.

❌ **The dataset is too small**

* A model with many independent variables requires a sufficient number of observations to avoid overfitting and ensure reliable predictions.

## Summary

| Condition                         | Multiple Linear Regression?  |
| --------------------------------- | ---------------------------- |
| Two or more independent variables | ✅ Yes                        |
| Continuous target variable        | ✅ Yes                        |
| Linear relationship               | ✅ Yes                        |
| Independent observations          | ✅ Yes                        |
| High multicollinearity            | ❌ Avoid or reduce it         |
| Non-linear relationship           | ❌ Consider non-linear models |
| Categorical target variable       | ❌ Use Classification         |

### Rule of Thumb

> **Use Multiple Linear Regression when a continuous target variable is influenced by two or more independent variables, and the relationships are approximately linear with minimal multicollinearity.**
