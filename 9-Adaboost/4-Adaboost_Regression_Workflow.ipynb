# AdaBoost Regression with Two Weak Learners

## 🏠 Example: Predicting House Prices Using Two Weak Learners

Suppose we want to predict the prices of houses using AdaBoost Regression.

### Dataset

| House | True Price |
| ----- | ---------: |
| A     |        200 |
| B     |        300 |
| C     |        400 |

---

## Step 1: Train the First Weak Learner

The first weak learner is a simple model that makes a rough prediction. Assume it predicts the **average house price** for every house.

The average price is:

[
\frac{200+300+400}{3}=300
]

Therefore, the predictions are:

| House | True Price | Prediction | Error |
| ----- | ---------: | ---------: | ----: |
| A     |        200 |        300 |  -100 |
| B     |        300 |        300 |     0 |
| C     |        400 |        300 |  +100 |

The first learner predicts the price of **House B** correctly but makes large errors for **Houses A and C**.

---

## Step 2: Focus on the Prediction Errors

AdaBoost gives **higher importance (weights)** to the samples with larger prediction errors.

Since Houses **A** and **C** have the largest errors, the next weak learner focuses more on improving their predictions.

The required corrections are:

| House | Required Correction |
| ----- | ------------------: |
| A     |                -100 |
| B     |                   0 |
| C     |                +100 |

---

## Step 3: Train the Second Weak Learner

The second weak learner learns to reduce the mistakes made by the first learner.

Suppose it predicts the required corrections accurately:

| House | Predicted Correction |
| ----- | -------------------: |
| A     |                 -100 |
| B     |                    0 |
| C     |                 +100 |

---

## Step 4: Combine the Two Weak Learners

The final prediction is obtained by combining the prediction of the first learner with the correction made by the second learner.

[
\text{Final Prediction}=\text{Prediction of Learner 1}+\text{Correction from Learner 2}
]

| House | Learner 1 | Learner 2 | Final Prediction |
| ----- | --------: | --------: | ---------------: |
| A     |       300 |      -100 |            200 ✅ |
| B     |       300 |         0 |            300 ✅ |
| C     |       300 |      +100 |            400 ✅ |

All three house prices are predicted correctly after combining the two weak learners.

---

## 🔑 Key Takeaways

* **Weak Learner 1** provides a rough estimate of the house prices.
* AdaBoost identifies the samples with larger prediction errors.
* **Weak Learner 2** focuses more on these difficult samples and learns to reduce the errors.
* The predictions from both learners are combined to produce a more accurate final prediction.
* By repeatedly adding weak learners, AdaBoost builds a strong regression model with improved prediction accuracy.

# AdaBoost Regression with Three Weak Learners

## 🏠 Example: Predicting House Prices Using Three Weak Learners

Suppose the true house prices are:

| House | True Price |
| ----- | ---------: |
| A     |        200 |
| B     |        300 |
| C     |        400 |

---

## Step 1: Train the First Weak Learner

The first weak learner predicts the average price for every house.

[
\frac{200+300+400}{3}=300
]

Predictions:

| House | Prediction | Error |
| ----- | ---------: | ----: |
| A     |        300 |  -100 |
| B     |        300 |     0 |
| C     |        300 |  +100 |

The model correctly predicts House **B**, but makes large errors for Houses **A** and **C**.

---

## Step 2: Train the Second Weak Learner

The second learner focuses on correcting the mistakes made by the first learner.

Suppose it predicts only **half of the required correction**.

| House | Predicted Correction |
| ----- | -------------------: |
| A     |                  -50 |
| B     |                    0 |
| C     |                  +50 |

---

## Step 3: Combine the First Two Learners

The updated predictions become:

| House | Calculation | Prediction |
| ----- | ----------- | ---------: |
| A     | 300 + (-50) |        250 |
| B     | 300 + 0     |        300 |
| C     | 300 + 50    |        350 |

Although the predictions are better, Houses **A** and **C** still have remaining errors.

---

## Step 4: Train the Third Weak Learner

AdaBoost again focuses on the remaining prediction errors.

The remaining corrections are:

| House | Remaining Correction |
| ----- | -------------------: |
| A     |                  -50 |
| B     |                    0 |
| C     |                  +50 |

Suppose the third learner again predicts only half of the remaining correction.

| House | Predicted Correction |
| ----- | -------------------: |
| A     |                  -25 |
| B     |                    0 |
| C     |                  +25 |

---

## Step 5: Combine All Three Weak Learners

The final predictions become:

| House | Calculation         | Final Prediction |
| ----- | ------------------- | ---------------: |
| A     | 300 + (-50) + (-25) |              225 |
| B     | 300 + 0 + 0         |              300 |
| C     | 300 + 50 + 25       |              375 |

The prediction errors are now much smaller than after the first learner.

---

## 🔑 Key Takeaways

* **Weak Learner 1** provides an initial rough estimate.
* **Weak Learner 2** reduces the large prediction errors.
* **Weak Learner 3** further improves the predictions by correcting the remaining errors.
* Each additional weak learner makes the overall model more accurate.
* This iterative process of **predict → identify errors → improve predictions → combine learners** allows boosting algorithms to build a strong regression model from multiple weak learners.
