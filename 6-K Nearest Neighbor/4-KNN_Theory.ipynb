# K-Nearest Neighbors (KNN)

## Introduction

**K-Nearest Neighbors (KNN)** is one of the simplest and most intuitive supervised machine learning algorithms. It is a **non-parametric**, **instance-based**, and **lazy learning** algorithm used for both **classification** and **regression** tasks.

Unlike many machine learning algorithms, KNN does not build a mathematical model during the training phase. Instead, it stores the entire training dataset and makes predictions only when a new data point is provided.

The main idea behind KNN is:

> **"Similar data points are likely to have similar outputs."**

For every new observation, the algorithm finds the **K closest data points (neighbors)** from the training dataset and predicts the output based on those neighbors.

---

# Real-Life Example

Imagine you move into a new neighborhood and want to know whether the area is considered "safe."

Instead of asking everyone, you ask your **5 nearest neighbors**.

* If 4 neighbors say the area is safe and 1 says it is not, you conclude that the area is probably safe.

This is exactly how KNN Classification works.

Similarly, if you want to estimate the average house price in the neighborhood, you can average the prices of the nearest houses.

This is how KNN Regression works.

---

# Types of KNN

KNN can solve two types of machine learning problems.

## 1. KNN Classifier

Used when the target variable is **categorical**.

Examples:

* Spam Detection
* Disease Prediction
* Loan Approval
* Customer Churn Prediction
* Handwritten Digit Recognition

Prediction Method:

* Finds the K nearest neighbors.
* Uses **majority voting**.
* Predicts the most common class.

---

## 2. KNN Regressor

Used when the target variable is **continuous**.

Examples:

* House Price Prediction
* Salary Prediction
* Temperature Prediction
* Sales Forecasting

Prediction Method:

* Finds the K nearest neighbors.
* Computes the average (or weighted average).
* Returns a numerical value.

---

# How KNN Works

The KNN algorithm follows these steps:

### Step 1: Store the Training Data

Unlike other algorithms, KNN does not create a mathematical model.

It simply stores all the training examples.

---

### Step 2: Choose the Value of K

K represents the number of nearest neighbors.

Examples:

* K = 3
* K = 5
* K = 7

Choosing the correct K is important because it directly affects model performance.

---

### Step 3: Calculate the Distance

When a new data point arrives, KNN calculates the distance between the new point and every training example.

The most common distance metric is **Euclidean Distance**.

[
d=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}
]

Other distance metrics include:

* Manhattan Distance
* Minkowski Distance
* Chebyshev Distance

---

### Step 4: Find the K Nearest Neighbors

Sort all distances in ascending order.

Select the first K nearest data points.

---

### Step 5: Make Prediction

For Classification:

* Perform majority voting.

For Regression:

* Compute the average of the neighbors' target values.

---

# Hyperparameters in KNN

## 1. n_neighbors (K)

Number of nearest neighbors.

* Small K → Sensitive to noise (Overfitting)
* Large K → Smoother predictions (May Underfit)

Common values:

* 3
* 5
* 7
* 9

---

## 2. weights

Determines how neighbors contribute.

Options:

* uniform
* distance

Uniform:
Every neighbor has equal importance.

Distance:
Closer neighbors receive higher importance.

---

## 3. algorithm

Algorithm used to search neighbors.

Options:

* auto
* ball_tree
* kd_tree
* brute

---

## 4. metric

Distance calculation method.

Common metrics:

* Euclidean
* Manhattan
* Minkowski
* Chebyshev

---

# Advantages of KNN

* Simple and easy to understand.
* Easy to implement.
* No training phase.
* Works for both classification and regression.
* Can model non-linear decision boundaries.
* Naturally supports multi-class classification.
* No assumptions about data distribution.

---

# Disadvantages of KNN

* Slow prediction for large datasets.
* Requires storing the entire dataset.
* Sensitive to irrelevant features.
* Sensitive to feature scaling.
* Performance decreases in high-dimensional datasets (Curse of Dimensionality).
* Choosing the correct value of K is difficult.
* Affected by noisy data and outliers.

---

# Time Complexity

### Training

Since KNN only stores data:

**O(1)**

---

### Prediction

Distance is calculated with every training sample.

**O(n × d)**

Where:

* n = Number of training samples
* d = Number of features

Prediction becomes slower as the dataset grows.

---

# Space Complexity

KNN stores all training samples.

**O(n × d)**

---

# Feature Scaling

KNN is a **distance-based algorithm**.

Features with larger values dominate the distance calculation.

Therefore, feature scaling is essential.

Common scaling methods:

* StandardScaler
* MinMaxScaler
* RobustScaler

---

# Choosing the Best Value of K

Very Small K

* Overfits
* Sensitive to noise
* High variance

Very Large K

* Underfits
* Ignores local patterns
* High bias

The optimal K is usually selected using **Cross-Validation**.

# When to Use KNN

Use KNN when:

* The dataset is small or medium-sized.
* Features are properly scaled.
* The relationship between features and target is non-linear.
* Similar observations have similar outputs.
* Model interpretability is important.

---

# When Not to Use KNN

Avoid KNN when:

* The dataset is very large.
* Fast predictions are required.
* The dataset has many irrelevant features.
* The data has very high dimensionality.
* The target contains many outliers (for regression).

---

# Interview Questions

### Why is KNN called a Lazy Learning Algorithm?

Because it does not learn a model during training. It stores the training data and performs computation only when making predictions.

---

### Why is KNN called a Non-Parametric Algorithm?

Because it does not assume any underlying distribution of the data and does not learn fixed model parameters.

---

### Why is Feature Scaling Important in KNN?

KNN relies on distance calculations. Features with larger scales can dominate the distance, leading to incorrect neighbor selection. Scaling ensures that all features contribute equally.

---

### Why Does KNN Become Slow for Large Datasets?

During prediction, KNN computes the distance between the new data point and every training sample. As the dataset size increases, the number of distance calculations also increases, making prediction slower.

---

# Summary

* It is used for both classification and regression.
* It predicts using the K nearest neighbors.
* Classification uses majority voting.
* Regression uses the average of neighbors.
* Feature scaling is essential.
* It performs well on small to medium-sized datasets but becomes slower as the dataset grows.
