# Top 10 Anomaly Detection Interview Questions and Answers

## 1. What is Anomaly Detection?

### Answer

**Anomaly Detection** is the process of identifying data points that are significantly different from the majority of the data.

These unusual observations are called:

* Outliers
* Anomalies
* Noise
* Rare Events

The main goal is to detect abnormal behavior that may indicate fraud, defects, failures, or suspicious activities.

**Example:**

In credit card transactions, if a customer usually spends ₹500–₹1000 but suddenly makes a ₹2,00,000 purchase from another country, it can be detected as an anomaly.

---

## 2. What are the different types of anomalies?

### Answer

There are **three main types** of anomalies:

### 1. Point Anomaly

A single observation that is very different from the rest.

**Example:**
One employee earning ₹10,00,000 while everyone else earns around ₹50,000.

---

### 2. Contextual Anomaly

A value is abnormal only in a specific context.

**Example:**
Temperature of **40°C** is normal in summer but abnormal during winter.

---

### 3. Collective Anomaly

A group of observations is abnormal even though individual observations may appear normal.

**Example:**
A sudden sequence of failed login attempts may indicate a cyber attack.

---

## 3. Which machine learning approaches are used for anomaly detection?

### Answer

### Supervised Learning

* Requires labeled normal and anomaly data.
* Used when anomaly labels are available.

### Unsupervised Learning

* Uses only unlabeled data.
* Assumes anomalies are rare.
* Most commonly used.

### Semi-Supervised Learning

* Trained using only normal data.
* Anything different from normal behavior is considered an anomaly.

---

## 4. Which algorithms are commonly used for anomaly detection?

### Answer

Some popular algorithms are:

* Isolation Forest
* DBSCAN
* Local Outlier Factor (LOF)
* One-Class SVM
* K-Means (Distance-Based)
* Gaussian Mixture Models (GMM)
* Autoencoders (Deep Learning)
* PCA (Principal Component Analysis)

---

## 5. Why is Isolation Forest popular for anomaly detection?

### Answer

Isolation Forest isolates anomalies instead of profiling normal data.

Since anomalies are rare and different, they are separated using **fewer random splits** than normal observations.

**Advantages**

* Very fast
* Handles large datasets
* No distance calculations
* Works well in high dimensions

---

## 6. How does DBSCAN detect anomalies?

### Answer

DBSCAN groups nearby data points into dense clusters.

Points that do not belong to any cluster are labeled as **Noise (-1)**.

These noise points are treated as anomalies.

---

## 7. What evaluation metrics are used for anomaly detection?

### Answer

Common evaluation metrics include:

### Precision

$$
\text{Precision}=\frac{TP}{TP+FP}
$$

Measures how many detected anomalies are actually anomalies.

---

### Recall

$$
\text{Recall}=\frac{TP}{TP+FN}
$$

Measures how many actual anomalies are successfully detected.

---

### F1-Score

$$
F1=\frac{2 \times Precision \times Recall}{Precision + Recall}
$$

Balances Precision and Recall.

---

### ROC-AUC

Measures how well the model separates normal and anomalous observations.

---

## 8. What challenges are faced in anomaly detection?

### Answer

Some common challenges are:

* Very few anomaly samples
* Imbalanced datasets
* Choosing the correct threshold
* High-dimensional data
* Noisy data
* False positives
* Changing data patterns over time (Concept Drift)

---

## 10. What is the difference between Outlier Detection and Anomaly Detection?

### Answer

| Outlier Detection                                    | Anomaly Detection                                                 |
| ---------------------------------------------------- | ----------------------------------------------------------------- |
| Focuses on unusual observations based on statistics. | Focuses on identifying abnormal behavior using machine learning.  |
| Usually analyzes individual data points.             | Can detect point, contextual, and collective anomalies.           |
| Mostly statistical methods.                          | Uses statistical, machine learning, and deep learning techniques. |

# Bonus Interview Questions

### 11. Why is anomaly detection considered an unsupervised learning problem?

Because anomaly labels are usually unavailable, and the model learns the normal data pattern to identify unusual observations.

---

### 12. What is a False Positive in anomaly detection?

A **False Positive** occurs when a normal observation is incorrectly classified as an anomaly.

---

### 13. What is a False Negative?

A **False Negative** occurs when an actual anomaly is incorrectly classified as normal.

---

### 14. Why are anomalies difficult to detect?

Because anomalies are:

* Rare
* Diverse
* Often hidden within normal data
* Continuously changing in many real-world applications

---

### 15. Which algorithm would you choose for anomaly detection?

It depends on the problem:

* **Isolation Forest** → Large, high-dimensional datasets
* **DBSCAN** → Spatial or density-based anomalies
* **One-Class SVM** → Small datasets with only normal training data
* **LOF** → Local density anomalies
* **Autoencoder** → Complex and high-dimensional data
* **PCA** → Detect anomalies through reconstruction error or reduced-dimensional representation

---

# Interview Tip

> **Remember this rule:**
>
> * **Isolation Forest** → Isolates anomalies using random splits.
> * **DBSCAN** → Detects anomalies as **Noise (-1)**.
> * **LOF** → Uses local density differences.
> * **One-Class SVM** → Learns the boundary of normal data.
> * **Autoencoder** → Uses reconstruction error to detect anomalies.
> * **PCA** → Finds anomalies based on projection or reconstruction error after dimensionality reduction.
