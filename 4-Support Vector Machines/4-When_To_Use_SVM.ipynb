# When to Use Support Vector Machine (SVM)

Support Vector Machine (SVM) is a supervised machine learning algorithm mainly used for **classification** problems. It is most effective when the data has a clear separation between classes and high prediction accuracy is required.

## Use SVM When

### 1. Small to Medium-Sized Dataset

* SVM performs well when the dataset is not extremely large.
* Example: **500 to 50,000 samples**.

### 2. High-Dimensional Data

* SVM works efficiently when the dataset has many features.
* Examples:

  * Text Classification
  * Email Spam Detection
  * Gene Expression Data

### 3. Clearly Separable Classes

* SVM is a good choice when the classes can be separated by a **line, plane, or hyperplane**.

### 4. High Classification Accuracy is Required

* SVM often provides excellent classification accuracy, especially for well-structured datasets.

### 5. Dataset Has a Clear Margin

* SVM finds the **Maximum Margin Hyperplane**, which improves the model's ability to generalize to unseen data.

### 6. Non-Linear Data

* If the data is not linearly separable, SVM can use **Kernel Functions** (such as **RBF** or **Polynomial Kernel**) to classify complex data.

# Example

**Problem:** Predict whether a tumor is **Benign** or **Malignant**.

**Features:**

* Tumor Radius
* Texture
* Area
* Perimeter

Since this is a **binary classification problem** with a **moderate-sized dataset**, **SVM is an excellent choice**.

---

# When NOT to Use SVM

* The dataset is **very large** (millions of samples), as training becomes slow.
* The dataset contains **a lot of noise** or **overlapping classes**.
* You need a **highly interpretable model** (Decision Tree or Logistic Regression is often better).
* The number of **samples is much larger than the number of features**, and simpler algorithms provide similar performance.

---

# Interview Question

### Q. When should you use SVM?

**Answer:**

Use **Support Vector Machine (SVM)** when the dataset is **small to medium-sized**, has **many features (high-dimensional data)**, and the classes are **clearly separable**. SVM provides **high classification accuracy** and can also classify **non-linear data** using **kernel functions** such as the **RBF** or **Polynomial Kernel**.
